(sec:coherence-quantum)=
# Quantum coherence functions

We extend the coherence functions defined with classical light in in {numref}`sec:classical-coherence` to quantum radiation.  The extension is straight forward once the radiation field is quantized. 

## Photons
Classical light is described by electromagnetic fields which are continuous both in space and time.  Quantum light is described also by electromagnetic fields but they are quantum fields.  The quantum fields are also continuous.  So, where is "quantumness" in the quantum light?  Photons immediately pops up in our head.  But what are they?  Unfortunately, they are poorly defined in literature  and misused by many people. In 1951 Albert Einstein said "All these fifty years of conscious brooding have brought me no nearer to the answer to the question 'What are light quanta?' Nowadays every Tom, Dick and Harry thinks he knows it, but he is mistaken."{cite:p}`French1979`  Since a correct use of the photon concept is so difficult, W. E. Lamb suggested that a license be required for use of the word photon."{cite:p}`Lamb1995`  Roy J. Glauber, who were given a license, defined photon as  “A photon is what a photodetector detects” and also "A photon is where the photodetector detects it" rather than defining a photon as a pre-existing "particle" in space.  When three Nobel laureates in quantum optics were saying like that, the meaning of photons must be very difficult for laymen to understand.

Although we don't have a license, we shall use the word "photon" in this note within a limited scope. We say either "A photon is emitted." or "A photon is detected."

If you are interested in the discussion of _What is a photon_, read Ref.{cite:p}`Roychoudhuri2017`.

## Quantization of electromagnetic fields

Once we have a set of modes (see {numref}`sec:coherence-opmodes`), field quatization is straight forward. The coefficients $a_{j}$ and $a^{*}_{j}$ become annihilation operator $\hat{a}_{j}$ and creation operators $\hat{a}^{\dagger}_{j}$, respectively. They satisfy usual commutation relation $[\hat{a}_{i},\hat{a}^{\dagger}_{j}] = \delta_{ij}$ and follow the usual operator algebra of a Harmonic oscillator.  The free Hamiltonian is given by $\sum_{j} \frac{\omega_{j}}{2} \left(a^{\dagger}_{j} a_{j} + \frac{1}{2}\right)$.
We define photons using the creation and annihilation operators. Defining a vacuum state by $a_{j}|0\rangle=0$, we create a photon in  $j$-th mode by $|1_j\rangle = a^{\dagger}_{j}|0\rangle$.  $N_{j} = a^\dagger_{j}a_{j}$ is a number operator as usual.  It is important to know that the operators are associated with the specific modes  we picked.  When different modes are used, we have different operators.   For example, consider a Bogoliubov transformation of basis from $u_{j}$ to $v_{k}$:

$$
v_{k} = \sum_{j} \left( \alpha_{kj} u_{j} + \beta_{kj} u_{j}^{*}\right).
$$

The annihilation operator for the new mode $v_{k}$ is given by

$$
\hat{b}_{k} = \sum_{j} \left( \alpha_{kj}^* \hat{a}_{j} - \beta_{kj}^* \hat{a}_{j}^{\dagger} \right).
$$

Note that the annihilation of photon in mode $v_{k}$ involves the creation of photon in mode $u_{j}$.  In general number of photons is not invariant under the transformation. This is quite confusing.  Are photons really  particles?

It is tempting to think that photons are particles like electrons.  They are NOT.  There is no position operator for photons, indicating that they are not point particles. Since the operators are tied to the modes, the number of photons also depends on the choice of modes.  The bottom line is that the quantum electromagnetic fields are fundamental but photons are emergent.  Some people call photon a pseudo particle.  This is why Glauber said "a photon is what a photodetector detects", which implies that we should not think about photons traveling around in the space. 

Until a license is given, we are  not allowed to discuss the details of photons in this note.   We use the word in two different occasions.  When a detector records a "click", we say a photon is detected.  When an electron in an emitter jumps from a higher energy state to a lower energy state, we say a photon is emitted.  What these statements mean is that the electron deposits a quanta of energy to the field, and then the energy stored in the field somehow travels to the detector and gives a click. That is all we need for now.



### Relation with photons and electric fields.

Suppose that the experimental setup can see only a single mode. 
Replacing , the electric field corresponding to the mode is expressed by

$$
\begin{eqnarray}
\mathbf{E}^{(+)}(\mathbf{x},t) &=& \mathbf{e} \mathcal{E} a(t) u(\mathbf{x}) \\
\mathbf{E}^{(-)}(\mathbf{x},t) &=& \mathbf{e} \mathcal{E} a^{\dagger}(t) u^{*}(\mathbf{x}) 
\end{eqnarray}
$$

where $a(t)$ and $a^{\dagger}(t)$ are annihilation and creation operators in the Heisenberg picture.     $\mathcal{E}$ is a proportional constant.  Since the photons are in a single mode, the direction of polarization is omitted.  Then, the intensity of the field is given by

$$
I(\mathbf{x},t) = \langle E^{(-)}(\mathbf{x},t) E^{(+)}(\mathbf{x},t)\rangle = \mathcal{E}^{2} |u(\mathbf{x})|^2 \langle a^{\dagger}(t) a(t) \rangle = \mathcal{E}^{2} |u(\mathbf{x})|^2 \text{tr} \left[n \rho(t)\right]
$$(eq:intensity_quantum)

where $n=a^{\dagger} a$ is the number operator and $\rho(t)$ is the density operator of photons. Th expectation value $\langle \cdots \rangle \equiv \text{tr} (\cdots \rho)$ is an ensemble average.  At the last step we switched  from the the Heisenberg picture to the Schr&ouml;dinger picture. 

Equation {eq}`eq:intensity_quantum` indicates that the intensity of light (in the mode we are looking at) is simply an average number of photons in the mode.    Be aware that we are counting only the photons in a particular mode, not necessarily all photons in the space. 

## Single Mode Detection

We begin with a simple case where the detector interacts with only one mode.  

### Quantum first-order coherence function

The classical first-order correlation function {eq}`eq:g1-classical-single-mode`  can be easily extended to a quantum system  by replacing $a^{*}$ with $a^{\dagger}$ and temporal average $\langle \cdots \rangle_{t}$ with average over a steady state $\langle \cdots \rangle_{ss}$.

$$
g^{(1)}(\tau;\mathbf{x}_1,\mathbf{x}_2) 
= \frac{\langle a^{\dagger}(t+\tau)a(t)\rangle_{ss}}{\langle a^{\dagger}(t)a(t)\rangle_{ss}} e^{i\phi(\mathbf{x}_1,\mathbf{x}_{2})}
$$(eq:g1-qm-sinmgle-mode)

where the expectation value is taken with the steady state. The equality {eq}`eq:nodelay-g1-classical-single-mode` holds also in the quantum regime.   Like the classical case, the spacial information is completely canceled out.   

While Eq. {eq}`eq:g1-qm-sinmgle-mode` is expressed in the Heisenberg picture, actual calculation is done in the Schr&ouml;dinger picture with a trick based on the quantum regression theorem, which we will discuss in {numref}`sec:openqm-qme-intro`.  

### Quantum second-order coherence function

The extension of the classical second-order coherence function{eq}`eq:g2-classical-single-mode`  to the quantum regime needs a special care due to a rule of quantum field theory known as normal order, which guarantees that the vacuum state does not have energy in it.

$$
g^{(2)}(\tau;\mathbf{x}_1,\mathbf{x}_2) = \frac{\langle a^{\dagger}(t) a^{\dagger}(t+\tau) a(t+\tau) a(t)\rangle_{ss}}{\langle a^{\dagger}(t+\tau)a(t+\tau) \rangle_{ss}\langle a^{\dagger}(t)a(t) \rangle_{ss}}
$$(eq:g2-qm-single-mode)

where $: \cdots :$ indicate normal order.  Notice the specific order of operators.  The annihilation operators must go the the right and act on the ket directory due to the normal ordering.  In addition, the operator must act in the chronological order according to the experimental procedure.  Since the system is in a steady state, the mean number of photons does not depend on time.

We are particularly interested in the value at $\tau=0$.  Using the commutation relation, $a^{\dagger}a^{\dagger} a a = a^{\dagger} a a^{\dagger} a - a^{\dagger} a = n^2-n$, we find

$$
g^{(2)}(0) = \frac{\langle n(n-1) \rangle_{ss}}{\langle n \rangle_{ss}^2} = \frac{\langle n^2 \rangle_{ss}}{\langle n \rangle_{ss}^2} - \frac{1}{\langle n \rangle_{ss}}
$$(eq:nodelay-g2-qm-single-mode)

Introducing the fluctuation of photon numbers $(\Delta n)^2 = \langle n^2 \rangle_{ss} - \langle n \rangle_{ss}^2$,  Eq.{eq}`eq:g2-qm-single-mode` becomes

$$
g^{(2)}(0) = 1 + \left(\frac{\Delta n}{\langle n \rangle_{ss}}\right)^2 - \frac{1}{\langle n \rangle_{ss}}
$$


This nodelay second-order coherence function looks similar to the classical version {eq}`eq:nodelay-g2-classical-single-mode` but actually quite different.  While the classical value is always above 1, we cannot say so for the quantum version.  For example, if $\langle n \rangle_{ss}=1$ and $\Delta n = 0$, then we have $g^{(2)}(0) = 0$.  That means if the detector sees only one photon at a time, the nodelay second-order coherence function vanishes.  This is purely a quantum effect known as photon antibunching. Finding the value of $g^{(2)}(0)$ is one of our goals in this note.

## Multi-mode detection


Apart from the normal ordering of operators and averaging over a steady state, the mathematical expression of quantum coherence functions are essentially the same as the classical ones [Eqs. {eq}`eq:g1-multimodes-general` and {eq}`eq:g2-multimodes-general`.]  Here, we just write them without derivation.

The first-order coherence function is given by

$$
g^{(1)}(\tau,\mathbf{x}) = \sum_{ij} w_{ij}(\mathbf{x}) g^{(1)}_{ij}(\tau).
$$

with 

$$
g^{(1)}_{ij}(\tau) = \frac{\langle a^{\dagger}_{i}(t+\tau)a_{j}(t)\rangle_t}{\langle a_{i}^{\dagger}(t)a_{j}(t)\rangle_{ss}} 
$$

and the second-order coherence functions by

$$
g^{(2)}(\tau;\mathbf{x}_1,\mathbf{x}_2) = \sum_{ij} \sum_{i'j'} w_{ij}(\mathbf{x}_1) w_{i'j'}(\mathbf{x}_2) g^{(2)}_{ij,i'j'}(\tau)
$$(eq:g2-qm-multimodes-general)

with

$$
g_{ij,i'j'}^{(2)}(\tau) = \frac{\langle a^{\dagger}_{i'}(t) a^{\dagger}_{i}(t+\tau) a_{j}(t+\tau) a_{j'}(t)\rangle_t}{\langle a^{\dagger}_{i}(t) a_{j}(t)\rangle_t \langle a^{\dagger}_{i'}(t) a_{j'}(t)\rangle_{ss}}
$$(eq:subg2-qm-multimodes-general)

The definition of $w_{ij}(\mathbf{x})$ remains the same.[See Eq. {eq}`eq:wij`.]

## Is $g^{(1)}$ measurable?

The first-order coherence function is a correlation of electric fields measured at two spacetime points. Is that something we can measure in a real world experiment?  Actually not. In quantum mechanics, electric field is not an observable. It is even not a self-adjoint operator and thus $g^{(1)}(\tau)$ is inherently complex. It is like measuring a wave function or a density operator. We know that neither is possible. Based on the principle of quantum mechanics, there is no way to measure $g^{(1)}(\tau)$! 

Then, why do we use it? It is a mathematical utility to investigate coherence. In fact, it is a powerful tool. Is there anyway to reconstruct it from experimental data?  Yes, there is. It is called [quantum tomography](https://en.wikipedia.org/wiki/Quantum_tomography).  You measure many different cases and if there is enough information in data, it is possible to _infer_ the function by fitting data to the theory. Essentially, we adjust parameters until all data can be fit to the theory.  It is a tedious process.

There is no such difficulty with the second-order coherence function since it is a correlation of intensities which can be measured at any time and any point.  It is also noted that quantum trajectory measures also intensity.  Thus we can evaluate $g^{(2)}(\tau)$ from the quantum trajectories but not $g^{(1)}(\tau)$.
