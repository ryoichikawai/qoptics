# QME: Analytical solution

In this section, we solve the quantum master equation analytically as much as possible and find a closed form expression of the second-order coherence.

## Quantum master equation

Based on the model introduced in {numref}`sec:TLS1-model`, the  time evolution of the state $\rho$ follows the quantum master equation

$$
\frac{d}{dt} \rho = -\frac{i}{\hbar} \left[H_\text{rwa},\rho\right]+ \mathcal{D}(\rho)
$$(eq:TSL1-qme)

The master equation is an ordinary differential equation (ODE) of the density operator.  It is convenient to write the density operator in a matrix form or a density matrix. and write down the quation of motion for each matrix elements. Using $|g\rangle$ and $|e\rangle$ as basis set, we introduce matrix elements: $\rho_{gg} = \langle g|\rho|g\rangle$, $\rho_{ee}=\langle e|\rho|e\rangle$, $\rho_{eg} = \langle e |\rho| g\rangle$, and $\rho_{ge} = \langle{g}|\rho|e\rangle$. The master equation {eq}`eq:TSL1-qme` is explicitely written as

$$
\begin{eqnarray}
		\frac{d}{dt}\rho_{gg} &=& - i \frac{\Omega}{2}(\rho_{eg}-\rho_{ge}) + \gamma_0(N+1)\rho_{ee} - \gamma_0 N \rho_{gg} \\
		\frac{d}{dt}\rho_{ee} &=& - i \frac{\Omega}{2}(\rho_{ge}-\rho_{eg}) - \gamma_0(N+1)\rho_{ee} + \gamma_0 N \rho_{gg}	\\
		\frac{d}{dt}\rho_{ge} &=& i \Delta \rho_{ge} -i \frac{\Omega}{2}(\rho_{ee}-\rho_{gg}) - \frac{1}{2}\gamma_0(2N+1) \rho_{ge} \\
		\frac{d}{dt}\rho_{eg} &=& -i \Delta \rho_{eg} - i \frac{\Omega}{2}(\rho_{gg}-\rho_{ee}) - \frac{1}{2}\gamma_0(2N+1) \rho_{eg}
\end{eqnarray}
$$

Using the normalization condition $\rho_{gg}+\rho_{ee}=1$ and the Hermitian condition $\rho_{ge} = \rho^{*}_{eg}$, we can further simplify the equations. Introducing $Z \equiv \rho_{ee}$, $X = \equiv \text{Re} \rho_{eg}$ and $Y \equiv \text{Im}\rho_{ge}$, the matrix elements are given by $\rho_{gg}=1-Z$, $\rho_{ee}=Z$, $\rho_{eg}=X+iY$, and $\rho_{ge}=X-iY$.  Hence, we need to find three real functions $X(t)$, $Y(t)$, and $Z(t)$.  Their equations of motion are

$$
\begin{eqnarray}
	\frac{d}{dt} X &=& -\frac{\gamma}{2} X - \Delta Y \\
    \frac{d}{dt} Y &=& -\frac{\gamma}{2} Y + \Delta X - \Omega Z + \frac{\Omega}{2} \\
    \frac{d}{dt} Z &=& - \gamma Z + \Omega Y + \frac{\gamma-\gamma_0}{2} 
\end{eqnarray}
$$

where $\gamma = \gamma_0 (2N+1)$. Further introducing 

$$
\mathbf{x} = 
	\begin{pmatrix}
		X \\ Y \\Z 
	\end{pmatrix},
\qquad
\mathbf{b} = 
	\begin{pmatrix}
		0 \\ \frac{\Omega}{2} \\\frac{\gamma-\gamma_0}{2} 
	\end{pmatrix}
\qquad
A = 
	\begin{pmatrix}
		-\frac{\gamma}{2} & - \Delta & 0 \\
		\Delta & - \frac{\gamma}{2} & -\Omega \\
		0 & \Omega & - \gamma
	\end{pmatrix}
$$

The master equation is expressed as simpe linear ODE

$$
	\frac{d}{dt} \mathbf{x} = A \mathbf{x} + \mathbf{b}
$$(eq:TLS1-ODEs)

whose solution is given by

$$
    \mathbf{x}(t) = \mathbf{x}^{ss} + e^{A t} [\mathbf{x}_{0}-\mathbf{x}^{ss} ]
$$(eq:TLS1-qme-solution)

where $\mathbf{x}_{ss}$ is the steady state solution.  The original form of the quantum master equation looks complicated but it is just a simple linear ODEs. 



## Steady state solution

Every solution to the quantum master equations asymptotically approach to a steady state regardless of the initial condition since the every eigenvalue of $A$ has a negative real part.  Since we are interested in the photon counting in a steady state, it is convenient to have an analytical expression of the steady state. Setting $\frac{d}{dt} \mathbf{x} = 0$ in {eq}`eq:TLS1-ODEs` and using the inverse matrix

$$
A^{-1} = 	\frac{1}{\gamma(\gamma^2+2\Omega^2+4\Delta^2)}
	\begin{bmatrix}
		-2(\gamma^2+2\Omega^2) & 4\gamma\Delta & -4 \Delta \Omega\\
		-4\gamma\Delta & -\gamma^2 & 2\gamma\Omega\\
		-4\Delta\Omega & - 2 \gamma\Omega & -(\gamma^2+4\Delta^2)
	\end{bmatrix}
$$

we obtain the steady state

$$
\begin{eqnarray}
	X^{ss} &=& -\frac{2 \Delta \gamma_0 \Omega}{\gamma(\gamma^2+2\Omega^2+4 \Delta^2)}\\
	Y^{ss} &=& \frac{\gamma_0 \Omega}{\gamma^2 + 2 \Omega^2 + 4 \Delta^2}\\
	Z^{ss} &=& \frac{1}{2} - \frac{(\gamma^2 + 4 \Delta^2) \gamma_0}{2\gamma\left(\gamma^2+2\Omega^2 + 4 \Delta^2\right)}
\end{eqnarray}
$$(eq:XYZ_ss)

Going back to the original variables, the steady state density matrix is found to be

$$
\begin{eqnarray}
\rho_{gg}^{ss} &=& \frac{1}{2} + \frac{(\gamma^2 + 4 \Delta^2) \gamma_0}{2\gamma\left(\gamma^2+2\Omega^2 + 4 \Delta^2\right)} \\
\rho_{ee}^{ss} &=& \frac{1}{2} - \frac{(\gamma^2 + 4 \Delta^2) \gamma_0}{2\gamma\left(\gamma^2+2\Omega^2 + 4 \Delta^2\right)} \\
\rho_{ge}^{ss} &=& -\frac{(2 \Delta + i \gamma) \gamma_0 \Omega }{\gamma(\gamma^2+2\Omega^2+4 \Delta^2)} \\
\rho_{eg}^{ss} &=& -\frac{(2 \Delta - i \gamma) \gamma_0 \Omega }{\gamma(\gamma^2+2\Omega^2+4 \Delta^2)} 
\end{eqnarray}
$$







Let us look at the solution for a special case. When driving is absent, the system is expected to relax to a thermal state.  To simulate it, let $\Omega=0$.  Then, the off-diagonal elements of the steady state density vanishes.  The probability of the system being in the excited state is 

$$
\rho_{ee}^{ss} = \frac{1}{2} \left(1 - \frac{\gamma_0}{\gamma} \right) = \frac{N}{2N+1} = \frac{e^{-\omega_0/T}}{1+e^{-\omega_0/T}}
$$

which is precisely the Boltzmann distribution.  Hence, the steady state is the thermal state.

## Time evolution

It is possible to derive the time-evolution of the state from Eq. {eq}`eq:TLS1-qme-solution`.  Let $\lambda_{i}$ is the $i$-th eigenvalue of $A$ and $\mathbf{u}_{i}$ is the corresponding normalized eigenvector.  Then, the solution {eq}`eq:TLS1-qme-solution` can be written as

$$
\mathbf{x}(t) = x^{ss} + \sum_{i=1}^{3} e^{\lambda_i t} \mathbf{u}_i ~[\mathbf{u}_i \cdot (\mathbf{x}_{0}-\mathbf{x}^{ss})]
$$(eq:TLS1-qme-solution-eigen)

The actual expression of the eigenvalues and eigenvectors are a bit messy. Here we show simple cases where analytical calculation is tractable.


### Thermalization

As discussed above, the system relaxes to a thermal state in the absence of driving ($\Omega=0$ and $\Delta=\omega_0$.   Actually, we already found the solution for this case in {numref}`sec:qme` using the Bloch equation. As a practice, we derive it again without using the Bloch vector. 

$$
\begin{eqnarray}
	X(t) &=&  e^{-\gamma t/2} \left[\cos(\omega_0 t) X_{0} - \sin(\omega_0t) Y_{0} \right]\\
	Y(t) &=& e^{-\gamma t/2} \left[ \sin(\omega_0 t) X_{0} + \cos(\omega_0t) Y_{0} \right]\\
	Z(t) &=& Z^{ss} + e^{-\gamma t} (Z_{0} - Z^{ss})
\end{eqnarray}
$$

where $X_{0}$, $Y_{0}$, and $Z_{0}$ are initial conditions. This solution is equivalent to the solution {eq}`eq:qme-bloch-solution`.

### Resonant fluorescence

Even when the tuning is perfect ($\Delta=0$), the analytical solution is very messy.  Here we solve it approximately under the strong driving condition ($\Omega \gg \gamma$).  Three eigenvalues of $A$ are

$$
\lambda_{1} = - \frac{\gamma}{2}, \quad
\lambda_{2}  \approx -\frac{3\gamma}{4} - \Omega i, \quad
\lambda_{3}  \approx -\frac{3\gamma}{4} + \Omega i
$$

and the corresponding eigenvectors after the approximation are

$$
\mathbf{u}_{1} = \begin{pmatrix} 1\\0\\0 \end{pmatrix}, \quad \mathbf{u}_{2} \approx \frac{1}{\sqrt{2}} \begin{pmatrix} 0\\-i\\1 \end{pmatrix}, \quad
\mathbf{u}_{3} \approx \frac{1}{\sqrt{2}} \begin{pmatrix} 0\\i\\1 \end{pmatrix}
$$

Note that they are orthonormal assuming $\left(\frac{\gamma}{\Omega}\right)^2$ is negligibly small.

For the sake of $g^{(2)}(\tau)$ calculation, let us assume that the system is initially in the ground state, that is $\mathbf{x}_{0}=0$. After taking strong driving approxatiomation, the steady state becomes 

$$
X^{ss} = 0, \quad Y^{ss}\approx 0, \quad Z^{ss}\approx \frac{1}{2}
$$

Now substituting all these approximated values into Eq. {eq}`eq:TLS1-qme-solution-eigen`, we obtain

$$
\begin{eqnarray}
X(t) &=& 0\\
Y(t) &=& e^{-\frac{3\gamma}{4} t} \sin(\Omega t) / 2\\
Z(t) &=& \frac{1}{2}\left(1 - e^{-\frac{3\gamma}{4} t} \cos(\Omega t) \right )
\end{eqnarray}
$$

The state relaxes to the steady state with decay rate $\frac{3\gamma}{4}$ and oscillates with the Rabi frequency $\Omega$.

## Second-order coherence

Once we obtain the time-evolution of state starting from the ground state, we can calculate the second-order coherence easily as discussed in {numref}`sec:qme`.   Using the above results, we obtain the second-order coherence under the strong driving as

$$
g^{(2)}(\tau) = \frac{Z(\tau)}{Z^{ss}} = 1 - e^{-\frac{3\gamma}{4} |t|} \cos(\Omega t)
$$(eq:TLS1-g2-approx)

which shows anti-bunching behavior at $\tau=0$. Comparing it with the simple thermalization case discussed in {numref}`sec:qme`,  the Rabi oscillation is clearly seen in addition to the change in the decay rate.  The following plot shows strong positive correlation near $\tau = 0$, suggesting strong bunching effect.

```python
import numpy as np
import matplotlib.pyplot as plt

Omega = 2.0
gamma = 0.2

time = np.linspace(0,30,300)
g2 = 1 - np.exp(-3*gamma/4*time)*np.cos(Omega*time)

plt.figure(figsize=(4,3))
plt.plot(time,g2,c='k')
plt.plot(-time,g2,c='k')
plt.axhline(y=1.0,c='r',linestyle='--')
plt.xlabel(r"$\tau$")
plt.ylabel(r"$g^{(2)}(\tau)$")
plt.show()
```

We need to use numerical methods to see the effect of weak driving and off-resonant cases, which comes in the next section.
