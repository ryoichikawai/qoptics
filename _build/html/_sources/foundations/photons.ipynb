# Photons and quantum radiation fields

## What is a photon?
Classical light is described by electromagnetic fields which are continuous both in space and time.  Quantum light is described also by electromagnetic fields but they are quantum fields.  The quantum fields are also continuous.  So, where is "quantumness" in the quantum light?  Photons immediately pops up in our head.  But what are they?  Unfortunately, they are poorly defined in literature  and misused by many people. In 1951 Albert Einstein said "All these fifty years of conscious brooding have brought me no nearer to the answer to the question 'What are light quanta?' Nowadays every Tom, Dick and Harry thinks he knows it, but he is mistaken."{cite:p}`French1979`  Since a correct use of the photon concept is so difficult, W. E. Lamb suggested that a license be required for use of the word photon."{cite:p}`Lamb1995`  Roy J. Glauber, who were given a license, defined photon as  “A photon is what a photodetector detects” and also "A photon is where the photodetector detects it" rather than defining a photon as a pre-existing "particle" in space.  When three Nobel laureates in quantum optics were saying like that, the meaning of photons must be very difficult for laymen to understand.

We should not imagine that a photon is traveling along its trajectory like a classical particle.  The position of a photon does not have clear meaning. No position operator can be defined for a photon and thus we cannot measure its position. Recall that for another quantum particle, electron, the position can be measured.  So the situation of photon is quite different from other quantum particles.  Nevertheless, we assume that the speed of a photon is the same as the speed of classical light $c$.[^1] 

Although we don't have a license, we shall use the word "photon" in this lecture note but only in a limited scope. We use the word in two occasions:  we say "A photon is emitted" when a quanta of energy is transferred from the emitter to the radiation field, and "A photon is detected" when the detector clicked (received a quanta of energy from the field).   The mathematical definition of photons will be given below.

If you are interested in the discussion of _What is a photon_, read Ref.{cite:p}`Roychoudhuri2017`.

[^1]:  Think how the speed of photon can be experimentally measured. Strictly speaking, we can only infer it.

## Classical radiation field

Classical radiation fields are described by electromagnetic waves. An electromagnetic wave is a solution to the Maxwell equations, which is a vector in the Hilbert space of infinite dimension.  Hence, the solution can be always  expressed as a linear combination of basis functions $\{u_{i}(x)\}$:

$$
\mathbf{E}(\mathbf{x},t) = \sum_{j} \mathbf{e}_{j} \left( \mathcal{E}_{j} a_{j}(t) u_{j}(\mathbf{x}) +  \mathcal{E}^{*}_{j} a^{*}_{j}(t) u^{*}_{j}(\mathbf{x})\right )
$$(eq:field_decomposition)

where $\mathcal{E}_{j}$ and $\mathbf{e}_{j}$ are a constant with the dimension of the electric field and a polarization vector associate with the $j$-th basis function. The coefficients $a_{j}$ represent  a particular solution.  In optics, a basis is called an _optical mode_.

In a free space,  the most popular choice of the basis is plane waves $u_{j}(\mathbf{x}) = e^{i\mathbf{k}_{j}\cdot \mathbf{x}}$ and correspondingly, we have $a_{j}(t) = a_{j} e^{-i \omega_{j} t}$ and $\mathcal{E}_{j} = i \sqrt{\frac{\hbar \omega_j}{2 \varepsilon_{0} V}}$. We have used discrete wave vectors in order to avoid mathematical complication but in the real world, we have to use a continuous spectrum. 

In classical theory, the coefficients $a_{j}(t)$ are treated as stochastic variables. The randomness accounts for the apparent loss of coherence.  It works quite well to reproduce experimental data but is not based on the first principles.

In principle, the choice of modes are arbitrary. We can pick modes that are convenient for the physical situation we are investigating. A problem is that a convenient choice of modes to describe a light emitter is not necessarily a good choice for a detector. It is very important to pick modes consistent to the experimental setup.

For mathematical convenience, we use the complex expression $\mathbf{E} = \mathbf{E}^{(+)} + \mathbf{E}^{(-)}$  defined by

$$
\mathbf{E}^{(+)}(\mathbf{x},t) = \sum_{j} \mathbf{e}_{j} \mathcal{E}_{j} a_{j}(t) u_{j}(\mathbf{x}), \qquad \mathbf{E}^{(-)}(\mathbf{x},t) = \sum_{j} \mathbf{e}_{j} \mathcal{E}^{*}_{j} a^{*}_{j}(t) u^{*}_{j}(\mathbf{x}).
$$(eq:classical-fields)

and the intensity of the light is defined by

$$
I(\mathbf{x}) = \langle \mathbf{E}^{(-)}(\mathbf{x},t) \cdot \mathbf{E}^{(+)}(\mathbf{x},t)\rangle = \sum_{i,j} \mathbf{e}_{i}\cdot\mathbf{e}_{j}\, \mathcal{E}_{i} \mathcal{E}^{*}_{j} \, u_{i}(\mathbf{x}) u^{*}_{j}(\mathbf{x})  \langle a^{*}_{i}(t) a_j(t)\rangle
$$(eq:classical-intensity)

where $\langle \cdots \rangle$ is an ensemble average of the probability distribution of the random phase. The ensemble average is usually replaced with temporal average in practical application.

## Photons

Quantizing radiation field {eq}`eq:classical-fields` is straight forward. The coefficients $a_{j}$ and $a^{*}_{j}$ become annihilation operator $\hat{a}_{j}$ and creation operators $\hat{a}^{\dagger}_{j}$, respectively. They satisfy usual commutation relation $[\hat{a}_{i},\hat{a}^{\dagger}_{j}] = \delta_{ij}$ and follow the usual operator algebra of a Harmonic oscillator. 
We define photons using the creation and annihilation operators. Defining a vacuum state by $a_{j}|0\rangle=0$, we create a photon in  $j$-th mode by $|1_j\rangle = a^{\dagger}_{j}|0\rangle$.  $n_{j} = a^\dagger_{j}a_{j}$ is a number operator as usual.
We can say "a photon is created or destroyed.  We can also count photons.  That is all we can say about photons.

The free Hamiltonian of photons is given by $\sum_{j} \frac{\omega_{j}}{2} \left(a^{\dagger}_{j} a_{j} + \frac{1}{2}\right)$ where $\hbar \omega_{j}$ is the energy of photon is the $j$-th mode.


Since the operators are tied to the modes, _the number of photons also depends on the choice of modes_.  The bottom line is that the quantum electromagnetic fields are fundamental but photons are emergent.  Some people call photon a pseudo particle.  This is why Glauber said "a photon is what a photodetector detects", which implies that we should not think about photons traveling around in the space. 

When an electron in an emitter jumps from a higher energy state to a lower energy state, we say a photon is emitted.  What these statements mean is that the electron deposits a quanta of energy to the field, and then the energy stored in the field somehow travels to the detector and gives a click. That is all we need for now. The photons reveal their existence only when the energy quanta comes in and out.


## Quantized radiation fields

Here we consider the radiation field in a free space. It is natural to expand the field with the plane wave basis set. Then, the electric field at position $\mathbf{x}$ and time $t$ is given by

$$
\mathbf{E}(\mathbf{x}) = \sum_{\ell} \mathbf{e}_{\ell} \left( \mathcal{E}_{\ell} a_{\ell} e^{i \mathbf{k}_{\ell} \cdot \mathbf{x}} +  \mathcal{E}^{*}_{\ell} a^{\dagger}_{\ell} e^{-i \mathbf{k}_{\ell} \cdot \mathbf{x}}\right )
$$(eq:quantum-free-field)

where $\mathcal{E}_{\ell} = i \sqrt{\frac{\hbar \omega_{\ell}}{2 \epsilon_{0} V}}$ and $\epsilon_{\ell}$ is the polarization vector for the $\ell$-th mode.

We also use another popular notation which splits the field to the two parts:

$$
\mathbf{E}^{(+)}(\mathbf{x}) = \sum_{\ell} \mathbf{e}_{\ell} \mathcal{E}_{\ell} a_{\ell} e^{i \mathbf{k}_{\ell} \cdot \mathbf{x}}, \qquad
\mathbf{E}^{(-)}(\mathbf{x}) = \sum_{\ell} \mathbf{e}_{\ell} \mathcal{E}^{*}_{\ell} a^{\dagger}_{\ell} e^{-i \mathbf{k}_{\ell} \cdot \mathbf{x}}.
$$(eq:quantum-free-field-pm)

Remember that the electric field $\mathbf{E}$ is a quantum mechanical operator.  If the Heisenberg picture is used, $a$ and $a^\dagger$ are time-dependent and thus $\mathbf{E}$ is time-dependent.  If the Schrödinger picture is used, then $\mathbf{E}$ does not depend on time.  We will choose a convenient picture depending on the situation.

The quantum expectation value of an observable $\hat{A}$ is defined as $\langle A(t) \rangle = \text{tr}(A(t) \rho)$ in the Heisenberg picture or $\text{tr}(A \rho(t))$ in the Schrödinger picture, where $\rho$ is a density operator. 


The mean intensity of light is given by

$$
I(\mathbf{x},t) = \langle \mathbf{E}^{(-)} \cdot\mathbf{E}^{(+)}\rangle = \sum_{i,j} \mathbf{e}_{i}\cdot\mathbf{e}_{j}\, \mathcal{E}^{*}_{i} \mathcal{E}_{j} \, e^{-i \mathbf{k}_{i}\cdot \mathbf{x}} e^{i \mathbf{k}_{j} \cdot \mathbf{x}}  \langle a^{\dagger}_{i}(t) a_{j}(t)\rangle
$$(eq:radiation-intensity)

This averaging replaces the ensemble average appeared in the classical intensity {eq}`eq:classical-intensity`.

In classical optics, the electric field is a measurable quantity.  In quantum field theory, the observation of electric field is quite different.  If the state of radiation field is an eigenstate of the number operator, the expectation value of the field vanishes, .i.e. $\langle \mathbf{E} \rangle=0$.  However, $\langle \mathbf{E}^2 \rangle > 0$ even when there is no photon. This is just the properties of a harmonic oscillator.

The classical-like electric field can be seen when the quantum state of the radiation is a _coherent state_, a mixture of different numbers of photons. The state of laser beam is an example. 

We also need to be careful about  $\mathbf{E}^{(\pm)}$.  They are not self-adjoint and thus not physical observables.  Measuring $\mathbf{E}^{(\pm)}$ is not possible.



## Measurement of radiation field

In classical optics, the intensity of light is often defined as 

$$
\begin{aligned}
I &= \langle \mathbf{E}\cdot \mathbf{E} \rangle \nonumber \\
&= 
\langle (\mathbf{E}^{(+)})^2 \rangle + \langle (\mathbf{E}^{(-)})^2 \rangle + \langle \mathbf{E}^{(+)}\cdot \mathbf{E}^{(-)} \rangle +  \langle \mathbf{E}^{(-)}\cdot \mathbf{E}^{(+)} \rangle.
\end{aligned}
$$(eq:intensity-fullterms)

Equation {eq}`eq:radiation-intensity` is only one of the four terms.  Are we missing some energy?  There are several reasons quantum optics uses Eq. {eq}`eq:radiation-intensity`.

**Infinite zero point energy**

Suppose that we have a vacuum state (no photon).  $\langle 0 | \mathbf{E}^{(+)}\cdot \mathbf{E}^{(-)} | 0  \rangle$ diverges due to the zero point energy of the infinite number of oscillators.  We want to remove this energy from the intensity of the radiation.

**Rapid oscillations**

Since $\mathbf{E}^{(+)} \propto e^{-i \omega t}$ and $\mathbf{E}^{(-)} \propto e^{i \omega t}$, the square terms in Eq. {eq}`eq:intensity-fullterms` are rapidly oscillating and the standard detector does not have a sufficient time resolution to see the oscillation.  The detector averages out the oscillation and thus these terms vanishes.  Since the actual detector can't see them, there is no reason to include them. This is just a practical reason. 

**What a detector sees**

A single photon detector records the arrival of individual photons.  Inside the detector, an electron gains energy from a photon by absorbing it.  Mathematically, this process is represented by $a^{\dagger}a$.  Only $\langle \mathbf{E}^{(-)}\cdot \mathbf{E}^{(+)} \rangle$ corresponds to the process in the detector.

We need to be careful about  $\mathbf{E}^{(\pm)}$.  They are not self-adjoint and thus not physical observable.  Measuring $\mathbf{E}^{(\pm)}$ is not possible.  Any observable quantity must be a product of the same number of $\mathbf{E}^{(+)}$ and $\mathbf{E}^{(-)}$. 

In classical optics, the electric field is a measurable quantity.  In quantum field theory, the observation of electric field is quite different.  If the state of radiation field is an eigenstate of the photon number operator, the expectation value of the field vanishes, i.e., $\langle \mathbf{E}\rangle=0$  despite $\langle \mathbf{E}^2 \rangle$ diverges even when there is no photon as we discussed above.

The classical-like electric field can be seen when photons are in a _coherent state_, a mixture of different numbers of photons. The state of laser beam is an example.  In this lecture note, we are dealing with a weak radiation field so that we can count the number of photons arriving the detector.
