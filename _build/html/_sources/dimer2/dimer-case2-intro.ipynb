# Dimer II: Coherent driving

In {numref}`ch:dimer-case1`, two emitters were completely incoherent.  In this section, the emitters are still separated by a distance with no direct interaction. However, they are coherently excited by a single laser beam.  By sharing the same driving source, the emitters estabilsh coherency and the spontaneous emission from them makes interference similar to double slit experiments. It was first observed by Eechmann _et al._ {cite:p}`Eichmann1993` and theoretically explained by Wang _et al._ {cite:p}`Wong1997`.  In turns out that more interesting  spatio-temporal coherence was theoretically predicted from the same system {cite:p}`Skornia2001,Schoen2001`. Following these theoretical prediction, the position-dependent bunchin/anti-bunching {cite:p}`Wolf2020` and superradiance/subradianece {cite:p}`Richter2023` are experimentally observed.  We will simulate their results.

Let us think what coherent driving means.  Suppose that the two emitters are excited by the driving field and the system is in  state $|ee\rangle \otimes |phi_{0}\rangle$.  Hwne the two emitters are incoherent, spontaneous emission takes place randomely.
