# The model

In {numref}`ch:dimer-case1`, two emitters were completely incoherent.  In this section, we again consider a pair of emitters with no interaction between them. However, they are coherently excited by a single laser beam.  By sharing the same driving source, the emitters establish coherency and the spontaneous emission from them makes interference patterns similar to the double slit experiment. It was first observed by Eechmann _et al._ {cite:p}`Eichmann1993` and theoretically explained by Wang _et al._ {cite:p}`Wong1997`.  In turns out that more interesting  spatio-temporal coherence was theoreticall predicted from the same system {cite:p}`Skornia2001,Schoen2001`. Following these theoretical prediction, the position-dependent bunchin/anti-bunching {cite:p}`Wolf2020` and superradiance/subradianece {cite:p}`Richter2023` are recently reported.  We will simulate their results.

## Model

Consider a pair of QDs located at $\mathbf{x}_{1}$ and $\mathbf{x}_{2}$, respectively. We use other position vectors as indicated in {numref}`ch:emitter`,  the relative position $\mathbf{r}_{12} = \mathbf{x}_{1}-\mathbf{x}_2$, and the position vectors with respect to the midpoint between the QDs $\mathbf{r}_{1} = \frac{1}{2}\mathbf{r}_{12}$ and $\mathbf{r}_{2} = -\frac{1}{2}\mathbf{r}_{12}$. 

The key difference from {numref}`ch:dimer-case1` is the presence of an external driving  by a laser beam. (See {numref}`ch:driving`.)  The intensity of the laser is assumed to be $\Omega < \omega_0$.  The system Hamiltonian is now given by

$$
H_{S} = \sum_{j}\frac{\Delta}{2} \sigma^{z}_{j}   + \frac{\Omega_{L}}{2}\left(e^{i \phi} \sigma^{+}_{1} + e^{-i \phi} \sigma^{-}_{1}\right) 
+ \frac{\Omega_{L}}{2}\left(e^{-i \phi} \sigma^{+}_{2} + e^{i \phi} \sigma^{-}_{2}\right)
$$

where $\phi = \frac{1}{2} \mathbf{k}_{L} \cdot \mathbf{r}_{12}$ and $\mathbf{k}_{L}$ is the wave vector of the laser.

Since the system is driven, we do not need thermal excitation and thus $N_{T}=0$. 

The Liouville super operator is now defined by

$$
\mathcal{L} \rho =& - i  \left[ H_{S},\rho \right] 
+ \frac{1}{2}\sum_{j} \gamma_{0}  \left(2 \sigma^{-}_{j} \rho \sigma^{+}_{j} - \sigma^{+}_{j}\sigma^{-}_{j} \rho - \rho \sigma^{+}_{j}\sigma^{-}_{j} \right) \\
$$(eq:dimer-case2-Liouville)

## Changing basis set

In Ref. {cite:p}, the spatial coherence is computed in the product basus.  However, a certain level of coherence between two QDs is expected and thus the product basis may not be the best choice.  On the other hand,  the distance between the dots is not small and there is non-negligible phase difference. Hence, the Dick basis is also not appropriate. Skornia _et al._{cite:p}`Skornia2001` introduced the following basis (we shall call it the Skornia basis.)

$$
|e\rangle = |ee\rangle, \qquad |s\rangle = \frac{1}{\sqrt{2}}\left( e^{-i \phi} |eg\rangle + e^{+i \phi}|ge\rangle\right), \qquad
|a\rangle = \frac{1}{\sqrt{2}}\left( e^{-i \phi} |eg\rangle - e^{i \phi}|ge\rangle\right), \qquad |g\rangle = |gg\rangle
$$(eq:skornia_basis)

where $\phi = \frac{1}{2} \mathbf{k}_{L} \cdot \mathbf{r}_{12}$.  $|s\rangle$ and $|a\rangle$ are symmetric and anti-symmteric , respectivelly, with respect to the swapping of the QDs. (Do not forget that the swapping changes the sign of phase from $\pm \phi$ to $\mp \phi$.) Since the global phase factor does not change the state, it is sufficient to consider only $0 \le \phi < \pi$.  When $\phi=0$, we have the Dick basis.

The master equation was analytically solved in this basis and the diagonal element of the steady state is found to be {cite:p}`Skornia2001`:

$$
\rho_{gg} = \frac{\left(\gamma^2 + \Omega^2\right)^2}{\left(\gamma^2 + 2 \Omega^2\right)^2}, \qquad
\rho_{ss} = \frac{\Omega^2\left(2 \gamma^2+\Omega^2\right)}{\left(\gamma^2 + 2 \Omega^2\right)^2}, \qquad
\rho_{aa} = \rho_{ee} = \frac{\Omega^4}{\left(\gamma^2 + 2 \Omega^2\right)^2}
$$(eq:rho_in_Skornia_basis)

It is important to note that the diagonal elements do not depend on the phase $\phi$ and it is uniquely determined by $\Omega/\gamma$.  

Accordingly, we need to change the jump operators. 

$$
\eta^{\pm}_{1} &= \frac{1}{\sqrt{2}}\left (e^{ \pm i \phi} \sigma^{\pm}_{1} + e^{\mp i \phi} \sigma^{\pm}_{2} \right), \qquad
\eta^{\pm}_{2} &= \frac{1}{\sqrt{2}}\left (e^{ \pm i \phi} \sigma^{\pm}_{1} - e^{\mp i \phi} \sigma^{\pm}_{2} \right) 
$$(eq:skornia_ops)

The relation between the old and new jump operators is

$$
e^{\pm i\phi} \sigma^{\pm}_{1} = \frac{1}{\sqrt{2}}\left(\eta^{\pm}_{1} + \eta^{\pm}_{2}\right), \qquad 
e^{\mp i\phi} \sigma^{\pm}_{2} = \frac{1}{\sqrt{2}}\left(\eta^{\pm}_{1} - \eta^{\pm}_{2}\right).
$$

The new operators represent the jumps $\eta^{+}_1 |g\rangle = |s\rangle$, $\eta^{+}_1 |s\rangle = |e\rangle$,  $\eta^{+}_2 |g\rangle = |a\rangle$, $\eta^{+}_2 |a\rangle = -|e\rangle$.  

With these jump operators,  the driving Hamiltonian becomes

$$
H_{L} = \frac{\Omega}{\sqrt{2}} \left(\eta^{+}_{1} + \eta^{-}_{1}\right)
$$

Notice that the second mode $\eta^{\pm}_{2}$ does not appear in the driving Hamiltonian and thus the driving activates only on channel 1: $|g\rangle \Leftrightarrow |s\rangle \Leftrightarrow |e\rangle$ and no driving acts on channel 2: $|g\rangle \nLeftrightarrow |a\rangle \nLeftrightarrow |e\rangle$.  Only spontaneous emission takes place in the second channel: $|g\rangle \Leftarrow |a\rangle \Leftarrow |e\rangle$.  (See {numref}`fig:dimer-transition-diagram-2`.)



We can solve the master equation in any basis and there is no need to rewrite it with the new operators. If we wish, we can do so. The dissipator remains in the same form with the new jump operators in place of the old one:

$$
\mathcal{D} \rho = \frac{1}{2}\sum_{j} \gamma_{0}  \left(2 \eta^{-}_{j} \rho \eta^{+}_{j} - \eta^{+}_{j}\eta^{-}_{j} \rho - \rho \eta^{+}_{j}\eta^{-}_{j} \right) .
$$
