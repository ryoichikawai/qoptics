# Jump method algorithm

## The basic idea of the jump method

In the standard quantum mechanics, a quantum state the time evolution of $|psi\rangle$ is determined by the Schrödinger equation $i\frac{d}{dt}|\psi\rangle = H |\psi\rangle$, which preserves the normalization.  The amount of information encoded in the state remains the same all time.  This is true as long as the Hamiltonian is self-adjoint, i.e., $H^{\dagger} = H$.  The evolution is continuous and smooth.  However, such a unitary evolution alone cannot describe the whole story of quantum evolution.  When measurement is performed on the state, it instantaneously jumps to one of eigenvectors of the observable.  No unitary evolution can describe quantum measurement process.

Quantum jumps sees in a quantum trajectory are essentially the same as such quantum measurement processes.  The interaction between emitters and radiation field can be interpreted as continuous measurement by the radiation field, which causes the jumps.  Hence, we can't use the standard  Schrödinger equation to find a quantum trajectory. WE need a non-unitary evolution.

Suppose that in absence of the radiation field, the emitter evolves unitarily with  Hamiltonian $H_{S}$.  When it interacts with the radiation field jump from state $|\alpha\rangle$ to $|\beta\rangle$. Such jump is mathematically described by an operator $C=\sqrt{\gamma} |\beta\rangle\langle \alpha|$ where $\gamma$ is a rate of jump$.  $C$ is called collapse operator.

Next, we consider an effective Hamiltonian

$$
H_\text{eff} = H_{S} - \frac{i}{2} C^{\dagger} C
$$

Because of the imaginary unit $i$ in front of the collapse operator, the effective Hamiltonian is not self-adjoint.  Nevertheless, we use this Hamiltonian in the Schrödinger equation

$$
i \frac{d}{dt} |\psi\rangle = H_\text{eff} |\psi\rangle
$$(eq:fake-shcroedinger)

Obviously, the norm is not conserved.  Some information is leaking out of the system. But total information must be constant.  So, the radiation field must be receiving the information, which is photon.  We know that emission happens randomly and instantaneously.  So, the loss of information must happen as a jump.

Now, we describe this scenario mathematically.  Integrating Eq. {eq}`eq:fake-unitary` for brief time period $\delta t$,

$$
|\psi(t+\delta t)\rangle \approx \left(1 - i H_\text{eff} \delta t\right) |\psi(t)\rangle
$$(eq:fake-unitary)

Assuming that the state is initially normalized ($\langle\psi(t)|\psi(t)\rangle = 1$), the loss of the normalization is
$$
\begin{aligned}
\delta p &\approx  1 - \langle\psi(t+\delta t)|\psi(t+\delta t)\rangle \\
&= i \delta t \langle \psi(t)| H_\text{eff} - H^{\dagger}_\text{eff}|\psi(t)\rangle  \\
&= \delta t  \langle \psi(t)|C^{\dagger} C|\psi(t)\rangle  \\
&= \delta t \gamma |\langle \alpha|\psi(t)\rangle |^2
\end{aligned}
$$

First of all, $\delta p$ is clearly non-negative.  If $|\psi(t)\rangle$ is orthogonal to $|\alpha\rangle$, $\delta p = 0$.  The norm conserves since no transition can happen.  On the opposite limit, if $\psi(t)\rangle = |\alpha\rangle$ and $\delta t \gamma=1$, then $\delta p = 1$, that means the norm is completely lost.  Since $\gamma$ is the rate of jump, if the initial state is $|\alpha\rangle$, it will jump to $|\beta\rangle$ within $\delta t = 1/\gamma$.  We note that $\delta p$ behaves like a probability.  If $\delta p =0$, then there is no jump.  If $\delta p = 1$, a jump certainly takes place.  Based on this observation, we postulate that a jump takes place during time period $\delta t$ with probability $\delta p$.

## Algorithm

In the above example, we used only one collapse operator.  We can include as many collapse operators as we like:

$$
C_{j} = \sqrt{\gamma_j} |\beta_{j}\rangle\langle \alpha_{j}|.
$$

Here is the summary of algorithm.

**Initial preparation**
1.  Set a time step size $\delta t$.  it should be smaller than any of $1/\gamma_{j}$.
2.  Set initial time $t_0 = 0$ and loop counter $k=0$.
3.  Prepare an initial state $\psi(0)\rangle$ at time $t=0$ at random. It should be normalized.

**Main loop begins**
1. Increment a loop counter: $k=k+1$.
1. Set a new time $t_{k} = t_{k-1}+\delta t$.

**Choose Schr&ouml;dinger step or jump step at random**
1.  Evaluate $\delta p_{j} = \delta t \langle \psi(t_{k-1})|C_{j}^{\dagger} C_{j}|\psi(t_{k-1})\rangle$ for each $i$.
2.  Calculate probabilities $\delta p = \sum_{j} \delta p_{j}$.
3.  Generate a uniform random number $r \in [0,1]$.  If $r > \delta p$, then go to Schr&ouml;dinger step.  Otherwise go to jump step.

**Schr&ouml;dinger step**
1.  Compute a new state $|\psi(t_{k})\rangle = \left( 1 - i H_\text{eff} \delta t \right) |\psi(t_{k-1})\rangle$.
2.  Normalize the new state.
3.  Set jump flag = 0
4.  Go to recording step.


**Jump step**
1.  Generate a uniform random number $r \in [0,1]$.  Pick a jump channel $j$ based on the random number. 
2.  Compute a new state $|\psi(t_{k})\rangle = C_{j} | \psi(t_{k-1} \rangle$.
3.  Normalize the new state.
4.  Set jump flag = the channel number
5.  Go to recording step.

**Recording step**
1.  Record current time
2.  Using the new state vector, evaluate expectation value of desired observables.  Record them.
3.  If jump flag > 0, then record jump time
4.  Go to the next cycle of the main loop


# Relation with the quantum master equation

Using the state vector $|psi(t)\rangle$ obtained by the above jump methods, the density operator evolves as

$$
\begin{aligned}
\frac{d}{dt} \rho(t) &= \left(\frac{d}{dt}|\psi(t)\rangle\right)\langle\psi(t)| + |\psi(t)\rangle\left(\frac{d}{dt} \langle \psi(t)|\right)\\
& = -i H_\text{eff} |\psi(t)\rangle\langle\psi(t)| + i |\psi(t)\rangle\langle\psi(t)| H^{\dagger}_\text{eff} \\
& = -i [H_{S},\rho(t)] - \frac{1}{2} C^{\dagger} C \rho(t) - \frac{1}{2} \rho(t) C^{\dagger} C
\end{aligned}
$$(eq:fake-Liouoville)

SInce this evolution does not include the jumps yet, the norm of $\rho$ does not conserve. Taking trace of Eq. {eq}`eq:fake-Liouoville`, we find that the density operator is losing the normalization by $\langle C^{\dagger} C \rangle$ per unit time.  Hence, we need to add a jump term which compensates the loss.

Compare  Eq. {eq}`eq:fake-Liouoville` with the standard form of a quantum master equation, $C \rho(t) C^{\dagger}$ is missing.  This is exactly the state created by the jump, $C|\psi\rangle\langle \psi|C^{\dagger}= \gamma |\langle\alpha|\psi\rangle|^2|\beta\rangle\langle \beta|$. If this is added to Eq. {eq}`eq:fake-Liouoville`, the loss of the normalization is fully recovered. In fact, we can see that the additional term adds the destination state $|\beta\rangle\langle \beta|$ with the rate $\gamma |\langle\alpha|\psi\rangle|^2|$.  This rate is reasonable since $\gamma$ is the decay rate of $|\alpha\rangle$ and $|\langle\alpha|\psi\rangle|^2|$ is the probability that the state is in $|\alpha\rangle$.

Now we conclude that the density operator in a standard quantum master equation includes both the Schrödinger like evolution with $H_\text{eff}$ and jumps and thus  
the jump method is equivalent to the quantum master equation.  Since the jump method simulates the evolution of a single system rather than the ensemble, it is often said that the jump method unravels the corresponding quantum master equation.
