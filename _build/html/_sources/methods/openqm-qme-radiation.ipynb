# Far-field operator and coherence functions

The quantum master equation gives us a mathematical method to calculate the state of the emitters but it discarded the information of the radiation field.  Hence, the quantum master equation alone cannot describe the light emitted by the emitters.  Only we learn from the maser equation is the state of the emitters.  We need a separate theory for the radiation field generated by the emitters.

## The radiation field generated by the emitters

Suppose that the radiation is emitted at time $t$ by the $j$-th emitter located at $\mathbf{x}_{j}$.  That means a transition $\sigma^{-}_{j}$ took place at $t$. The detector is located at $\mathbf{x}_{d}$.  Let introduce a relative coordinate to the center of the quantum dot cluster. The detector position is $\mathbf{R} = \mathbf{x}_d - \frac{1}{N}\sum_{j}\mathbf{x}_{j}$ and the location of $i$-th dot is $\mathbf{r}_i = \mathbf{x}_{i} - \frac{1}{N}\sum_{j}\mathbf{x}_{j}$. We want to know the electric field that the detector interacts with at time $t_R = t + R/c$. 
The net radiation field is a sum of a free field $\mathbf{E}_{f}$ and a field generated by the source $\mathbf{E}_{s}$.  The free field is just the initial radiation field ({eq}`eq:free-radiation-field`).


Lehmberg {cite:p}`Lehmberg1970a` derived the source field by solving 
the Heisenberg equation of motion for $a_{\ell}(t)$ with the Born-Markov approximation, which is the same approximation used to derive the quantum master equation. The result is

$$
\mathbf{E}^{(+)}_{s}(\mathbf{R},t_{R}) = - \frac{\omega_{0}^{2} d}{c^2} \sum_{j} \frac{ \left(\hat{\mathbf{d}} \times \hat{\mathbf{R}}_{j}\right) \times \hat{\mathbf{R}}_{j}}{R_{j}} \, e^{i k_0(R_{j}-R)} \sigma^{-}_{j}(t)
$$(eq:source_field)

where $k_{0} = \omega_{0}/c$ and $\mathbf{R}_{j} = \mathbf{R} - \mathbf{r}_{j} = \mathbf{x}_{d} - \mathbf{x}_{j}$.  



Since the detector is far from the emitter, we can assume $|\mathbf{r}_{ij}| \ll \mathbf{R}_{j}|$, then Eq. {eq}`eq:source_field` can be approximated to

$$
\mathbf{E}^{(+)}_{s}(\mathbf{R},t_{R}) \approx  - \hat{\mathbf{e}}_{R} \frac{\omega_{0}^{2}d}{c^2} \left [ 1 - \left(\hat{\mathbf{d}}\cdot\hat{\mathbf{R}}\right)^2\right]^{1/2} \sum_{j} e^{- i k_{0} \hat{\mathbf{R}} \cdot \mathbf{r}_{j}} \sigma^{-}_{j}(t)
$$(eq:source_field_approx)

where $\mathbf{e}_{R}$ is a unit vector in the direction of $\left(\mathbf{d} \times \mathbf{R} \right)\times \mathbf{R}$.  Notice that the field strength depends on the direction of the radial vector through the two inner products $\hat{\mathbf{d}} \cdot \hat{\mathbf{R}}$ and $\hat{\mathbf{R}} \cdot \mathbf{r}_{j}$.  The latter depends the geometric structure of the emitters.  Now the source field is expressed with the operators $\sigma^{-}_{j}$ and $\sigma^{+}_{j}$ instead of $a_{\ell}$ and $a^{\dagger}_{\ell}$.  This allows us to evaluate the coherence functions with $\sigma^{-}_{j}$ and $\sigma^{+}_{j}$ which we can obtain from the quantum master equation.  

The detector sees both free and source fields.  The signal from the free field is the background noise to the source signal. It is not possible to eliminate the noise.  We assume that the noise from the free field is removed by some experimental process.

## Detection of photons

In {numref}`ch:detector`, we learned the Glauber's theory of detection.  A history of clicks recorded by a detector corresponds to the measurement of radiation intensity.  However, a quantum master equation does not have that information.  Folloiwng Lehmberg {cite:p}`Lehmberg1970a`,  we can make a connection between the density matrix and intensity through the source radiation field  {eq}`eq:source_field_approx`.    The intensity at the position of detector $\mathbf{R}$ is given by

$$
I(\mathbf{x},t) = \langle E^{(-)}(\mathbf{x},t)E^{(+)}(\mathbf{x},t)\rangle \rangle =  
\langle J^{+}(\mathbf{R},t) J^{-}(\mathbf{R},t) \rangle
$$


where

$$
J^{\pm} (\mathbf{R},t) = \sqrt{\gamma(\mathbf{R})} \sum_{i} e^{ \pm i k_{0} \hat{\mathbf{R}} \cdot \mathbf{r}_{j}} \sigma^{\pm}_{j}(t).
$$(eq:jump-operator-R)

with the transition rate 

$$
\gamma(\mathbf{R}) = \left(\frac{\omega_{0}^{2}d}{c^2}\right)^{2} \left [ 1 - \left(\hat{\mathbf{d}}\cdot\hat{\mathbf{R}}\right)^2\right].
$$

Recall that in the jump methods, we assume that the state vector jumps to a new state determined by the collapse operator.  Is $J^{-}(\mathbr{R})$ a collapse operator? The radiation field alone cannot determine it because the field does not know if the light from each emitter is coherent to others. It is determined by the other side, the emitters. (See {numref}`ch:coherence`.)  Since we know the state of the emitters, we can evaluate the expectation value of the intensity and determine the coherency:

$$
\begin{aligned}
I(\mathbf{x},t) &= \langle J^{+}(\mathbf{R},t) J^{-}(\mathbf{R},t) \rangle \\
&= \left(\frac{\omega_{0}^{2}d}{c^2}\right)^2 \left [ 1 - \left(\hat{\mathbf{d}}\cdot\hat{\mathbf{R}}\right)^2\right] \sum_{ij} e^{i k_0 \hat{\mathbf{R}}\cdot \mathbf{r}_{ij}} \langle \sigma^{+}_{i} \sigma^{-}_{j}\rangle \\
\end{aligned}
$$

If $\langle \sigma^{+}_{i} \sigma_{j}\rangle = 0$ for $i\ne j$, then the light from an emitter is not coherent with others.
In this case, $\sqrt{\gamma(\mathbf{R})} \sigma^{-}_{j}$ is the collapse operator. In other limit, $J^{-}(\mathbf{R})$ is the collapse operator. 


It is important note that we are talking about the collapse operator at the detector (we shall call it far-field collapse operator. This is only a small part of the collapse operators for the emitters.  To connect the far-field operators to the collapse operators in the quantum master equation, we need to count all photons emitted in all direction.  Integrating $\gamma(\mathbf{R})$ over the solid angle

$$
\int_{0}^{2\pi} \int_{-\pi}^{\pi} \gamma(\mathbf{R}) \sin(\theta) d\theta d\phi = \gamma_0
$$

which is the standard rate of spontaneous emission and we obtain the collapse operators $\sqrt{\gamma_0} \sigma^{-}_{j}$.   

## Coherence function

We now measure the second-order coherence function by a detector placed at $\mathbf{R}$.  Using the far-field operators,

$$
g^{(2)}(\tau;\mathbf{R}) 
= \frac{\langle J^{+}(\mathbf{R},t) J^{+}(\mathbf{R},t+\tau) J^{-}(\mathbf{R},t+\tau) J^{-}(\mathbf{R},t)  \rangle_{ss}}{(\langle J^{+}(\mathbf{R},t) J^{-}(\mathbf{R},t)\rangle_{ss})^2} =
= \frac{\sum_{ij} \sum_{k\ell} e^{i k_0 \hat{\mathbf{R}}\cdot \mathbf{r}_{ij}}e^{i k_0 \hat{\mathbf{R}}\cdot \mathbf{r}_{k\ell}} \langle\sigma^{+}_{i}(t) \sigma^{+}_{k}(t+\tau)\sigma^{-}_{\ell}(t+\tau)\sigma^{-}_{j}(t)\rangle_{ss}}{ \left( \sum_{ij} e^{i k_0 \hat{\mathbf{R}}\cdot \mathbf{r}_{ij}}\langle\sigma^{+}_{i}(t)\sigma^{-}_{j}(t)\rangle_{ss}\right)^2 }.
$$(eq:g2-src-field)

For the incoherent light, we use  $\sqrt{\gamma} \sigma^{-}_{j}$ as jump operators and we obtain

$$
g^{(2)}(\tau;\mathbf{R}) 
= \frac{\sum_{ij}  \langle\sigma^{+}_{i}(t) \sigma^{+}_{j}(t+\tau)\sigma^{-}_{j}(t+\tau)\sigma^{-}_{i}(t)\rangle_{ss}}{ \left( \sum_{j} \langle\sigma^{+}_{j}(t)\sigma^{-}_{j}(t)\rangle_{ss}\right)^2 }.
$$(eq:g2-src-field-incoherent)

which does not depend on $\mathbf{R}$.

## Quantum regression theorem

There is a problem with the definitions {eq}`eq:g1-emission` and {eq}`eq:g1-emission`.  We don't how to calculate the expectation values since two different times are involved. There is a trick that works only when the Born-Markovian approximation is valid as discussed next.

Suppose that the expectation value of an operator $A$ evolves in time. It should be written as $\langle A \rangle_t$ (expectation evaluated at time $t$) but  often written as $\langle A(t) \rangle$ which suggests that $A$ depends on time. If the system evolution is unitary, then $A(t)$ can be considered as an operator in the Heisenberg picture and the average is taken with respect to an initial state.  However, when the evolution of the system is not unitary, we can't define the usual Heisenberg picture. We interpret it as $\langle A(t) \rangle \equiv \text{tr}(A \rho(t))$  where the state is evolving in time and the operator is stationary.

A problem arises when operators depend on two different times.  Fortunately, there is a very usefull theorem, known as the quantum regression theorem, which states that

$$
\langle A(\tau)B(0) \rangle = \text{tr} [ A e^{\mathcal{L}t} (B \rho(0)) ]
$$

Notice that neither $A$ nor $B$ depends on time in the right hand side of the equality. $B \rho(0)$ is evaluated at $t=0$ and evolves to t=$\tau$.  Then we take the trace. It is a kind of surprising that the same evolution operator $e^{\mathcal{L}t}$ can be used to evolve $B\rho$.  We can calculate $e^{\mathcal{L} t} (B \rho(0))$ by solving the quantum master equation with the initial condition  $\eta(0) = B\rho(0)$.


As an example, the four-operator correlation in {eq}`(eq:g2-src-field` can be computed as

$$
\begin{aligned}
\langle\sigma^{+}_{i}(0) \sigma^{+}_{k}(\tau)\sigma^{-}_{\ell}(\tau)\sigma^{-}_{j}(0)\rangle_{ss} &=
\text{tr}\left[\sigma^{+}_{i}(0) \sigma^{+}_{k}(\tau)\sigma^{-}_{\ell}(\tau)\sigma^{-}_{j}(0) \rho_{ss}\right] \\
&= \text{tr}\left[\sigma^{+}_{k}(\tau)\sigma^{-}_{\ell}(\tau)\sigma^{-}_{j}(0) \rho_{ss} \sigma^{+}_{i}(0)\right]\\
& = \text{tr}\left[\sigma^{+}_{k}\sigma^{-}_{\ell} e^{\mathcal{L}\tau} \left(\sigma^{-}_{j} \rho_{ss} \sigma^{+}_{i}\right)\right]
\end{aligned}
$$

where we started with the Heisenberg picture and switched to the Schrodinger picture at the end.
