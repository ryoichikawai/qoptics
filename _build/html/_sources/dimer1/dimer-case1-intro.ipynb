# Dimer I: Incoherent emitters 

In {numref}`ch:coherence`, we discussed the origin of decoherence.  When the radiation field can't distinguish the emitters ($\langle \phi_{1} | \phi_{1}^\prime\rangle \approx 1$), coherence is observed in the subsystem.  Otherwise ($\langle \phi_{1} |\phi_{1}^\prime\rangle \ll 1$), we don't see coherence.   The field very close to the emitters (near field) can obviously see them.   So, when the radiation field failed to see individual atoms.  The field far from the emitters (far field) is expressed by Eq.{eq}`eq:source_field_approx`.  The difference between the emitters appears in the phase factor, 
$e^{- i k_{0} \hat{\mathbf{R}} \cdot (\mathbf{r}_{1}-\mathbf{r}_{2})}$.   Roughly speaking, if $r_{12} k_{0} \ll 1$, then the phase difference vanishes and the far field can't distinguish the emitters.

In this chapter, we investigate the coherence properties of thermal fluorescence emitted from a pair of emitters(we shall call them a dimer) using a simple model. 

See Ref. {cite:p}`Ficek2002` for a comprehensive review of two-atom systems.
