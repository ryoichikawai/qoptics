# Radiation field

We have already define the free radiation field in {numref}`ch:photons`. A brief summary is given here again.  The key point in this chapter is the interaction between the emitter and the field.

## Photon gas

The free radiation field can have a finite temperature.  We call it a photon gas. The free Hamiltonian of photon is defined by

$$
H_{R} = \sum_{\ell} \omega_{\ell} \left (a^{\dagger}_{\ell} a_{\ell} + \frac{1}{2} \right)
$$

where $a_{\ell}$ and $a_{\ell}^{\dagger}$ are annihilation and creation operators of the $\ell$-th mode with momentum $\mathbf{k}_{\ell}$ respectively ($|\mathbf{k}_{\ell}| = \omega_{\ell}/c$).  We will omit the zero point energy in the rest of this note.

Initially the photon gas is assumed to be in a thermal equilibrium at temperature $T$.  The average occupation number follows the Planck distribution (Boltzman constant $k_{B}=1$ is assumed)

$$
\langle a_{\ell}^{\dagger} a_{\ell}\rangle_{0} = \frac{1}{e^{\omega_{\ell}/T} - 1}.
$$

When $T=0$, $\langle a_{\ell}^{\dagger} a_{\ell}\rangle_{0} =0$ for all modes.

When the photon gas interacts with the emitter, eventually the whole system reaches a new thermal equilibrium unless the emitter is externally driven.

## Free radiation field


We have defined the free radiation field as (See {eq}`eq:quantum-free-field-pm`.)
$$
\mathbf{E}^{(+)}(\mathbf{x}) = \sum_{\ell} \mathbf{e}_{\ell} \mathcal{E}_{\ell} a_{\ell} e^{i \mathbf{k}_{\ell} \cdot \mathbf{x}}, \qquad
\mathbf{E}^{(-)}(\mathbf{x}) = \sum_{\ell} \mathbf{e}_{\ell} \mathcal{E}^{*}_{\ell} a^{\dagger}_{\ell} e^{-i \mathbf{k}_{\ell} \cdot \mathbf{x}}.
$$

where $a^{\dagger}_{\ell}$ and $a_{\ell}$ are creation and annihilation operators of photons in the plane wave mode $e^{i \mathbf{k}_{\ell} \cdot \mathbf{x}}$ with linear polarization $\epsilon_{\ell}$ and $\mathcal{E}_{\ell} = i \sqrt{\frac{\hbar \omega_{\ell}}{2 \epsilon_{0} V}}$.

## Interaction between emitter and radiation

The interaction between the atoms and the radiation field is simply due to the force exerted by the radiation field on the electron.  Assuming that the wave length of light emitted by the atom is much larger than the size of the atom itself, electric dipole approximation can be used and the interaction Hamiltonian can be written as

$$
H_{SR} = -\sum_{j}\mathbf{d}_j \cdot \mathbf{E}(\mathbf{x}_{j}) 
$$(eq:SR-coupling)

where  $\mathbf{d}_{j}$ is the dipole moment operator of the $j$-th atom and $\mathbf{x}_{j}$ is the center position of the atom. The dipole moment operator is defined by

$$
\mathbf{d}_{j} = \mathbf{d}  \sigma^{+}_{j} + \mathbf{d}^{*} \sigma^{-}_{j}
$$(eq:dipole-moment-op)

with  $\mathbf{d} = - \langle e| \mathbf{p} |g\rangle$ where $\mathbf{p}$ is the electric dipole moment of the atom, which is assumed to be identical to all the atoms. 

Substituting Eq. {eq}`eq:quantum-free-field-pm`, we have a full expression of the interaction Hamiltonian

$$
H_{SR} = - \sum_{j} \left[\mathbf{E}^{(+)}(\mathbf{x}_{j})+\mathbf{E}^{(-)}(\mathbf{x}_j)\right] \cdot \left( \mathbf{d}  \sigma^{+}_{j} + \mathbf{d}^{*} \sigma^{-}_{j} \right)
$$(eq:H_SR)

Recall that $E^{(-)}$ creates a photon and $E^{(+)}$ destroys one, and that $\sigma^{+}$ and $\sigma^{-}$ represents excitation and deexcitation, respectively.  Thenm $E^{(-)} \sigma^{-}$ and $E^{(+)} \sigma^{+}$ describe ordinary emission and absorption processes.  However, Eq. {eq}`eq:H_SR` contains two more terms, $E^{(-)} \sigma^{+}$ and $E^{(+)} \sigma^{-}$ which do not make a sense from the perspective of classical physics.  In quantum mechanics, emission and excitation can simultaneously happens despite the process does not conserve energy.  These processes take place during very short period of time and the Heisenberg uncertainty principle allows non-energy conserving process as long as the energy is conserved in a longer time scale. 

These virtual processes play key roles in various optical processes such as [spontaneous parametric down-conversion](https://en.wikipedia.org/wiki/Spontaneous_parametric_down-conversion) and [Raman scattering](https://en.wikipedia.org/wiki/Raman_scattering).  Their controibution to regular optical processes is small and we ignore them under the name of [rotating wave approximation](https://en.wikipedia.org/wiki/Rotating-wave_approximation).

## Lamb shift and dipole-diploe interaction

The above interaction generates spontaneous emission and absorption of photons. When the emitted photon is detected by a detector, the actual light is observed.  There are other processes through the above interaction.  Individual atoms can emit and absorb virtual photons.  The atoms are said to be dressed with photons. A consequence is the shift of energy, known as the Lamb shift. Since the experimental value of the energy includes the Lamb shift, we will not add it here.  

Another virtual processes is the exchange of photons between two atoms.  Within the dipole approximation, it becomes the dipole-dipole interaction between atoms introduced in {numref}`ch:emitter`.  Higher-order interaction is not considered in this note.
