# Emitters

The emitter of photons is a group of identical atoms (or quantum dots).Particularly we are interested in a small cluster of atoms, the number of atoms ranging from $N=1$ through $N=5$ or so.  We focus on a particular transition between two energy levels and ignore other levels.  Mathematically, the two-level system (TLS) can be treated as a qubit (or a system of spin-1/2).

## A single two-level system

A single atom has two energy levels, a lower energy state $|g\rangle$ and a higher energy state $|e\rangle$ with corresponding energy $E_{g} < E_{e}$.  We ignore electron spin. Such a system is called two-level system (TLS).  $|g\rangle$ and $|e\rangle$ form an orthonormal basis set in the two-dimensional Hilbert space. We can consider the system as a qubit and the Pauli operators are defined by

$$
\sigma^{x} = | e \rangle\langle g | + | g \rangle\langle e |, \qquad \sigma^{y} =  i (| e \rangle\langle g | -  | g \rangle\langle e |), \qquad \sigma^{z} = | e \rangle\langle e | - | g \rangle\langle g |
$$

and raising and lowering operators by

$$
\sigma^{+} = \frac{1}{2}\left(\sigma^{x} - i \sigma^{y}\right) = | e \rangle\langle g |, \qquad \sigma^{-} =  \frac{1}{2}\left(\sigma^{x} + i \sigma^{y}\right) = | g \rangle\langle e |.
$$

Recall how the raising and lowering operators work:

$$
\sigma^{+}|g\rangle = |e\rangle, \quad \sigma^{+}|e\rangle =0, \qquad
\sigma^{-}|g\rangle = 0, \quad \sigma^{-}|e\rangle = |g\rangle
$$

For numerical calculation, the basis vectors $|e\rangle$ and $|g\rangle$ are expressed in the following matrix form

$$
|e\rangle = \begin{pmatrix} 1 \\ 0\end{pmatrix}, \qquad |g\rangle = \begin{pmatrix} 0 \\ 1\end{pmatrix}.
$$

Be careful about this definition since some people define them in in the opposite way.

Using the these notations, a free Hamiltonian for a single quantum dot is defined by

$$
H_{S} = \frac{\omega_0}{2} \sigma^{z}
$$

where $\omega_0 = E_{e}-E_{g}$  ($\hbar=1$ is assumed.)


## $N$ identical atoms

We assume that all atoms are identically modeled as TLS, with same excitation energy $\omega_0$ and the same dipole moment $d$.  Only the difference is their location.

   The dimension of the Hilbert space is $2^N$. For computational purpose, we use the standard product basis set,

$$
|s_{1} s_{2} \cdots s_{N}\rangle = |s_{1}\rangle\otimes|s_{2}\rangle\otimes \cdots \otimes |s_{N}\rangle, \qquad |s_{j}\rangle = |g\rangle \text{ or } |e\rangle
$$(eq:product-basis-N)

where $\otimes$ denotes tensor product. For $N=2$, we have 

$$
|gg\rangle = |g\rangle \otimes |g\rangle, \quad |ge\rangle = |g\rangle \otimes |e\rangle, \quad |eg\rangle = |e\rangle \otimes |g\rangle, \quad |ee\rangle=|e\rangle \otimes |e\rangle.
$$(eq:product-basis-2)

<p>The spin operator for the $j$-th atom is given by

$$
\sigma^{\mu}_{j} = \overbrace{I \otimes \cdots \otimes  I}^{j-1} \otimes \sigma^{\mu} \otimes I \otimes \cdots \otimes I, \qquad \mu = {x,y,z,\pm} 
$$

where $I$ is the identity operator for the two-dimensional Hilbert space.   For N=2. $\sigma^{\mu}_{1} = \sigma^{\mu} \otimes I$ and $\sigma^{\mu}_{2} = I \otimes \sigma^{\mu}$.

Then, the free Hamiltonian of the $N$ atoms is a simple sum

$$
H_{0} = \sum_{j=1}^{N} \frac{\omega_0}{2} \sigma^{z}_{j}
$$(eq:atom-free-hamiltonian)

where $\sigma^{z}_{j}$ is a Pauli operator for $j$-th quantum dot.


## Location of atoms

We denote the position of the $j$-th atom as $\mathbf{x}_{j}$. The relative position $\mathbf{r}_{ij} = \mathbf{x}_{i} - \mathbf{x}_{j}$ actually takes more important roles.

The location of atoms does not show up in the free Hamiltonian.  It will appear in the interaction Hamiltonian with radiation fields which will be discussed in {numref}`ch:radiation-field`.  The dipole-dipole interaction introduced below is the part of atom-radiation interaction and depends on the relative position.



## Dipole moment

Each atom carries a dipole moment $\mathbf{d}$.  Although we don't have to make a specific assumption, for mathematical simplicity it is assumed to be a linear dipole moment, which is a real vector. 

$$
\mathbf{d} = -e \langle e | \mathbf{r}_{e} | g \rangle
$$

where $\mathbf{r}_{e}$ is the position of electron with respect to the atom center and $-e$ is the electron charge.

## Interaction between the atoms

There are many possible interactions between atoms.  When the wave functions of two neighboring atoms overlap, quantum tunneling, exchange interaction and [the Dexter energy transfer](https://en.wikipedia.org/wiki/Dexter_electron_transfer) are possible.  Even in the absence of the overlap, the Förster Resonance Energy Transfer (FRET) and inter-atoms Auger processes can take places.  For quantum dots, the exchange of phonons between the dots can destroy quantum phase (dephasing).

In the present lecture note, we consider only the dipole-dipole interaction which is an indirect interaction through the free radiation field. The Hamiltonian for the dipole-dipole interaction is given by

$$
H_{dd} = \sum_{i \ne j} \Omega_{ij} \sigma^{+}_{i} \sigma^{-}_{j}
$$


with the strength of the dipole-dipole interaction {cite:p}`Lehmberg1970a` 

$$
\Omega_{ij} = \frac{3}{4} \gamma_{0} \left \{- \left[1-\frac{\left|\mathbf{d}\cdot\mathbf{r}\right|^2}{|d|^2|r_{12}|^2}\right] \frac{\cos \xi_{ij}}{\xi_{ij}} + \left[1 - 3\frac{\left|\mathbf{d}\cdot\mathbf{r}\right|^2}{|d|^2|r_{12}|^2}\right]\left(\frac{\sin \xi_{ij}}{\xi_{ij}^2}+\frac{\cos\xi_{ij}}{\xi_{ij}^3}\right)\right\}.
$$(eq:dd-cupling-strength)

where $\gamma_{0} = \frac{4 \omega_{0}^3 d^2}{3 \hbar c^3}$ is the rate of spontaneous emission and $\xi_{ij}=\omega_0 r_{ij}/c$. 

If the dipole is linear, $\frac{\left|\mathbf{d}\cdot\mathbf{r}\right|^2}{|d|^2|r_{12}|^2} = \cos^2\theta$ where $\theta$ is the angle between $\mathbf{d}$ and $\mathbf{r}_{12}$.
