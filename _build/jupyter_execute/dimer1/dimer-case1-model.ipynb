# Dimer I: Model

## System

Each emitter is modeled by a two-level system as defined in {numref}`sec:emitter`.  For computational purpose, we use the standard product basis set $|gg\rangle$, $|ge\rangle$, $|eg\rangle$, and $|ee\rangle$ (see {numref}`sec:emitter` for the definition of the product base.)

The four-dimensional spin operators are defined by (see {numref}`sec:emitter`)

$$
\sigma_{1}^{\mu} = \sigma^{\mu} \otimes I, \qquad \sigma_{2}^{\mu} = I \otimes \sigma^{\mu} , \qquad \mu = x,y,z, \pm.
$$

where $I$ is the identity operator in the two-dimensional Hilbert space.

The Hamiltonian of the free system is then expressed as

$$
H_{0} = \sum_{j=1}^{2} \frac{\omega_0}{2} \sigma^{z}_{j}
$$(eq:dimer-case1-hamiltonian)

We assume that there is no interaction between the emitters.

## Dissipator

As discussed in {numref}`sec:openqm-qme-method`, the definition of the disppator is ambiguous.  We need to pick collapse operators consistent with the physical situation.   When the emitters are well separated, the radiation field interacts with them individually.  That means jumps take places in each emitter independently.  Hence, there are four jumps.  The collapse operators    for absorption and emission by the $j$-th emitter are $\sqrt{\gamma_{a}} \sigma^{+}_{j}$ and $\sqrt{\gamma_{e}} \sigma^{-}_{j}$, respectively.

Accordingly, the dissipator for the pair is given by

$$
\begin{aligned}
\mathcal{D}(\rho) 
&=  \frac{1}{2} \gamma_{e} \left(2 \sigma^{-}_{1} \rho \sigma^{+}_{1} - \sigma^{+}_{1}\sigma^{-}_{1} \rho - \rho \sigma^{+}_{1}\sigma^{-}_{1} \right) \\
&+  \frac{1}{2} \gamma_{e} \left(2 \sigma^{-}_{2} \rho \sigma^{+}_{2} - \sigma^{+}_{2}\sigma^{-}_{2} \rho - \rho \sigma^{+}_{2}\sigma^{-}_{2} \right) \\
&=  \frac{1}{2} \gamma_{a} \left(2 \sigma^{+}_{1} \rho \sigma^{-}_{1} - \sigma^{-}_{1}\sigma^{+}_{1} \rho - \rho \sigma^{-}_{1}\sigma^{+}_{1} \right) \\
&+  \frac{1}{2} \gamma_{a} \left(2 \sigma^{+}_{2} \rho \sigma^{-}_{2} - \sigma^{-}_{2}\sigma^{+}_{2} \rho - \rho \sigma^{-}_{2}\sigma^{+}_{2} \right)
\end{aligned}
$$(eq:dimer-case1-dissipator)

where the emission and absorption rates are defined $\gamma_{e} = \gamma_{0} (N_{T}+1)$ and $\gamma_{a} = \gamma_{0} N_{T}$.  See {numref}`sec:openqm-qme-example` for details.

There is no off-diagonal dissipator since $\sigma^{\pm}_{1}$ and $\sigma^{\pm}_{2}$ are not correlated.

## Coherence function

We now measure the second-order coherence function by a detector placed at $\mathbf{R}$. Using the source field {eq}`eq:g2-src-field` and discard the cross correlation between the emitters, we obtain

$$
g^{(2)}(\tau;\mathbf{R}) 
= \frac{\sum_{ij} \langle\sigma^{+}_{i}(t) \sigma^{+}_{j}(t+\tau)\sigma^{-}_{j}(t+\tau)\sigma^{-}_{i}(t)\rangle_{ss}}{ \left( \sum_{i} \langle\sigma^{+}_{i}(t)\sigma^{-}_{i}(t)\rangle_{ss}\right)^2}.
$$(eq:g2-dimer-incoherent)

The position dependency vanishes since the radiations from the two emitters are incoherent (no interference).
