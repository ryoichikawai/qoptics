# Dimer I: Incoherent emitters 

We investigate thermal fluorescence emitted by an incoherent pair of emitters.  We assume that the pair, which we shall call a dimer, are completely independent to each other and thus there is no coherence in the product basis (Recall that coherence depends on the choice of the basis set, See {numref}`ch:coherence`.) . Let us look at the entanglement more carefully. 

Consider the product basis:

$$
|gg\rangle = |g\rangle \otimes |g\rangle, \quad |ge\rangle = |g\rangle \otimes |e\rangle, \quad |eg\rangle = |e\rangle \otimes |g\rangle, \quad |ee\rangle = |e\rangle \otimes |e\rangle
$$

Suppose that the system is $|Psi\rangle =  c_0(t) |ee\rangle \otimes |\phi_{0}\rangle + c_{1}(t) |ge\rangle \otimes |\phi_{1}\rangle + c_{1}^\prime(t)  |eg\rangle \otimes |\phi_{1}^\prime\rangle + c_{2} |gg\rangle \otimes |\phi_{2}\rangle$ where $|\phi_{n}\rangle$ indicates the state of field with $n$ photon.   $|\phi_{1}\rangle$ and $|\phi_{1}^\prime\rangle$ are two different states of the radiation field containing one photon.  In order to see the coherence in this basis, we calculate the reduced density.  In particular we are interested in $\rho_{eg} = c_{1} (c_{1}^\prime)^{*} |eg\rangle\langle ge| \langle \phi_{1}^\prime|\phi_{1}\rangle$.   If $\langle \phi_{1}^\prime|\phi_{1}\rangle = 0$, there is no coherence between $|eg\rangle$ and $|ge\rangle$. (See {numref}`ch:coherence`.)   In the opposite end,  if $\langle \phi_{1}^\prime|\phi_{1}\rangle = 1$, they are perfectly coherent (superradiance).  Othrewise, they are partially coherent.

In this chapter, we assume that $\langle \phi_{1}^\prime|\phi_{1}\rangle = 0$ so that $|eg\rangle$ and $|ge\rangle$ are incoherent.


We first introduce a model the model and then investigate the coherence properties using the quantum master equation and unravel the processes using the jump method.
