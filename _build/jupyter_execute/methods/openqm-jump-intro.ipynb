# Jump method and Quantum Trajectory

The quantum master equation  is useful if we are interested in the expectation value of observables.   However, it provides the state of an ensemble but no information about individual processes.  Recall that when there is no coherency, the off-diagonal elements of density matrix vanish and thus there is no transition between two states on average. It does not mean that the transition is not happening.  There are fluctuations and some individuals in the ensemble may briefly deviate from the average and allow the transition.
Even when the density operator of a system is in a steady state, the individual members of the ensemble are not stationary.  Their state jumps from one to another at random.  The time sequence of jumps is called _quantum trajectory_.  

Shcrödinger did not like the idea of quantum jumps. {cite:p}  Once he said:


>    "If we have to go on with these damned quantum jumps, then I'm sorry that I ever got involved"  
>    (E. Shcrödinger)


Zeh, an expert of quantum mechanics and coherence theory, did not like it either. {cite:p}`Zeh1993`

How do we know that quantum jumps happen?    It seems that no quantum trajectory of two-level system has been experimentally observed, perhaps because the process is too rapid.  Despite the difficulties, quantum trajectories have been observed in a two-level system with additional staging state where jumps between the excited state and the staging state were clearly seen.{cite:p}`Nagourney1986,Bergquist1986,Sauter1986`  See also a review of the measurement method {cite:p}`Minev2019`.  

Now that we know quantum jumps exist. How can we calculate it?  Not with a quantum master equation.  We need an alternative method suitable for monitoring individual processes. 
Such method is also better for full counting statistics.   

In this chapter, we introduce the jump method in which the system jumps from a state to another and produces quantum trajectories.  The jump method is becoming popular in quantum optics research and there are plenty of literature about it.  Here are some:
