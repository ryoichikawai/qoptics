# Brief summary of a quantum master equations

## Derivation of quantum master equation - a summary

Here a very brief summary of the derivation steps is provided so that we know what approximations are used.  Read a textbook listed in the previous page for the full derivation.
Consider a system S surrounded by a thermal reservoir R.  We assume that the reservoir is an ideal heat bath which remains in thermal equilibrium all time.  which is a key assumption to derive quantum master equation. The total system S+R is completely isolated and thus its time evolution is unitary.  The Hamiltonian of the total system can be divided to three parts:

$$
H_\text{S+R} = H_\text{S} \otimes I_\text{R} + I_\text{S} \otimes H_\text{R} + V_\text{SR}
$$

where $H_\text{S}$,  $H_\text{R}$ and $V_\text{SR}$ are the Hamiltonian for the system S, the reservoir R, and the interaction between S and R.  $I_\text{S}$ and $I_\text{R}$ are identity operators in each Hilbert space. 

The state of the total system is described by the density operator $\rho_\text{SR}$ and its time evolution is determined by the Liouville-von Neumann equation

$$
\frac{d}{dt} \rho_\text{SR} = -i \left[H_\text{SR}, \rho_\text{SR}\right].
$$

To solve this equation, two approximations are introduced.

1. The interaction between S and R is weak enough to use the Born approximation (1st order perturbation). Only $V_\text{SR}$ and ($V_\text{SR})^2$ are taken into account and all higher order terms are neglected.
2. The reservoir is always in the thermal state $\rho_\text{R} = \text{tr}_\text{S} \left( \rho_\text{SR} \right)= \frac{1}{Z_\text{R}} e^{-\beta H_\text{R}}$ where $\text{tr}_\text{S}$ is a partial trace over the system Hilbert space.  $\beta$ and $Z_\text{R}=\text{tr}_\text{R} e^{-\beta H_\text{R}}$ are usual inverse temperature and the partition function.
3. When the reservoir is disturbed, it _instantaneously_ relaxes to the original equilibrium state.  Hence, the state of reservoir remains in $\rho_\text{R}$ all time.  This implies that the reservoir has no memory of past events. This assumption is called the Markovian approximation. 
4. The coupling Hamiltonian is separable.  Mathematically, $V_\text{SR} = \sum_{i} A_i \otimes B_i$.  Further we assume that $\langle B_i \rangle = \text{tr} \left( B_{i} \rho_\text{R} \right) = 0$.
5. We introduce jump operators, each of which represents either excitation or deexcitation process.  

   $$
   A_{i} (\omega) = \sum_{k,k'} \delta(E_{k} - E_{k'} - \omega) |k\rangle \langle k | A_{i} |k' \rangle \langle k'|
   $$(eq:qme-jump-op)

   $A_{i}(\omega)$ and $A_{i}^{\dagger}(\omega)$ behave like annihilation and creation operators for $i$-th mode. From the completeness relation $\sum_{\omega} A_{i}(\omega) = \sum_{\omega} A_{i}^{\dagger}(\omega) = A_i$.
6. All the effects of the reservoir are taken into account through the temporal correlation of the reservoir operators.

   $$
   c_{ij}(\tau) = \text{tr}_\text{R} \left(B_j(\tau) B_i(0) \rho_\text{R}\right), \quad \tau>0.
   $$


7. Based on the Born-Markovian approximation, the state of the system can be approximated as $\rho_\text{SR} (t) \approx \rho_\text{S} (t) \otimes \rho_\text{R}$ where $\rho_\text{S} = \text{tr}_\text{R} \rho_\text{SR}$. 

At last we have a general form of quantum master equation.

$$
\frac{d}{dt} \rho_\text{S} = -i \left[ H_\text{S} + H_\text{LS}, \rho_\text{S}\right] + \mathcal{D} \rho_\text{S} 
$$(eq:qme_general)

where $\mathcal{D}$ is a superoperator known as dissipator.  It acts on a density operator as

$$
\mathcal{D} \rho  = \sum_{\omega} \sum_{ij} \Gamma_{ij}(\omega) \left[ A_{j}(\omega) \rho A_{i}^{\dagger}(\omega) - \frac{1}{2} \left\{ A_{i}^{\dagger}(\omega)A_{j}(\omega), \rho \right\} \right]
$$(eq:dissipator_general)

where $\{\cdot,\cdot\}$ is anti-commutator and the dissipation rate $\Gamma_{ij}$ is the Fourier transform of the correlation function:

$$
\Gamma_{ij} (\omega) = \int_{-\infty}^{+\infty} c_{ij}(\tau) e^{i\omega \tau} d\tau.
$$

In Equation {eq}`eq:qme_general`, the unitary part include is a Lamb shift $H_\text{LS}$ defined by

$$
H_\text{LS} = \frac{1}{2i} \sum_{\omega} \sum_{ij}  \left(\Gamma_{ij}(\omega) - \Gamma_{ji}^{*}(\omega)\right)
$$

Mathematically speaking, the Lamb shift diverges.  To resolve this issue, we need to use mass renormalization in the relativistic quantum field theory.  The shift is found to be very tiny.  Hence, we shall omit it throughout this lecture note.

The quantum master equation {eq}`eq:qme_general` works well only when the Born-Markovian approximation is valid.  In many applications in quantum optics, the approximation holds reasonably well.

To simplify mathematical notation, the master equation {eq}`eq:qme_general` is often expressed as

$$
\frac{d}{dt} \rho = \mathcal{L} \rho
$$

with  a super operator  $\mathcal{L}$, known as Liouvillian, which acts on density operator as

$$
\mathcal{L} \rho = -i \left[ H_\text{S} + H_\text{LS}, \rho \right] + \mathcal{D} \rho
$$

which allow us to write a solution to the master equation as

$$
\rho(t) = e^{\mathcal{L} t} \rho(0).
$$

We shall utilize the simplistic expression from time to time.

By changing the basis, the $\Gamma_{ij}$ in Eq. {eq}`eq:dissipator_general` can be diagonalized.  The diagonal form of the disspator is written 

$$
\mathcal{D} \rho  = \sum_{\omega} \sum_{i} \Gamma_{i}(\omega) \left[ L_{i}(\omega) \rho L_{i}^{\dagger}(\omega) - \frac{1}{2} \left\{ L_{i}^{\dagger}(\omega)L_{i}(\omega), \rho \right\} \right]
$$

which is the standard form shown in literature.  The operator $L_i$ is often called Lindblad operator or _jump operator_.  Yet another form is often used, in particular QuTiP.  The $\Gamma_i$ is absorbed into the definition of jump operators and the dissipator is written as  

$$
\mathcal{D} \rho  = \sum_{\omega} \sum_{i} \left[ C_{i}(\omega) \rho C_{i}^{\dagger}(\omega) - \frac{1}{2} \left\{ C_{i}^{\dagger}(\omega)C_{i}(\omega), \rho \right\} \right]
$$

where $C_{i}(\omega) = \sqrt{\Gamma_i(\omega)} L_i(\omega)$, which we shall call _collapse operator_.

## Quantum regression theorem

There is a problem with the definitions {eq}`eq:g1-emission` and {eq}`eq:g1-emission`.  We don't how to calculate the expectation values since two different times are involved. There is a trick that works only when the Born-Markovian approximation is valid as discussed next.

Suppose that the expectation value of an operator $A$ evolves in time. It should be written as $\langle A \rangle_t$ (expectation evaluated at time $t$) but  often written as $\langle A(t) \rangle$ which suggests that $A$ depends on time. If the system evolution is unitary, then $A(t)$ can be considered as an operator in the Heisenberg picture and the average is taken with respect to an initial state.  However, when the evolution of the system is not unitary, we can't define the usual Heisenberg picture. We interpret it as $\langle A(t) \rangle \equiv \text{tr}(A \rho(t))$  where the state is evolving in time and the operator is stationary.

A problem arises when operators depend on two different times.  Fortunately, there is a very usefull theorem, known as the quantum regression theorem, which states that

$$
\langle A(\tau),B(0) \rangle = \text{tr} [ A e^{\mathcal{L}t} (B \rho(0)) ]
$$

Notice that neither $A$ nor $B$ depends on time in the right hand side of the equality. $B \rho(0)$ is evaluated at $t=0$ and evolves to t=$\tau$.  Then we take the trace. It is a kind of surprising that the same evolution operator $e{\mathcal{L}t}$ can be used to evolve $B\rho$.  We can calculate $e^{\mathcal{L} t} (B \rho(0))$ by solving the quantum master equation with the initial condition  $\eta(0) = B\rho(0)$.

## Ambiguity

The expression of a quantum master equation is not unique.  It is invariant under many transformation.  Suppose that the dissipator is given by

$$
\mathcal{D}\rho = \sum_{j} \left(A_{j} \rho A^{\dagger}_{j} - \frac{1}{2} \left \{A^{\dagger}_{j}A_{j},\rho\right\}\right)
$$

Introducing a set of new operators using a unitary transformation $U$ such that $B_{j} = \sum_{k} U_{jk} A_{k}$, the dispator is transformed to

$$
\mathcal{D}\rho = \sum_{j} \left(B_{j} \rho B^{\dagger}_{j}- \frac{1}{2} \left \{B^{\dagger}_{j}B_{j},\rho\right\}\right)
$$

Even under a simple shift $B_{j} = A_{j} + \alpha_{j}$ where $\alpha_{j}$ is a constant, the dissipator is invariant.  Hence, there are infinitely many expression of the same quantum master equation.  


**Example**

Consider a pair of emitters interacting with the free radiation field.  The quantum master equation for the emitters is

$$
\frac{d}{dt} \rho = -i [H_{S},\rho] + \mathcal{D}\rho
$$

with the free Hamiltonian

$$
H_{S} = \sum_{j} \frac{\omega_0}{2} \sigma^{z}_{j}
$$

and

$$
\mathcal{D} \rho = \gamma \sum_{j} \left( \sigma^{+}_{j} \rho \sigma^{-}_{j} - \frac{1}{2} \left\{\sigma^{-}_{j}\sigma^{+}_{j}, \rho\right\} \right)
$$

Now introducing new operators

$$
\eta^{\pm}_{1} = \sigma^{\pm}_{1} + \sigma^{\pm}_{2}, \qquad \eta^{\pm}_{2} = \sigma^{\pm}_{1} - \sigma^{\pm}_{2}
$$

This transformation does not change $H_{S}$.  The original dissipator is transformed to 

$$
\mathcal{D} \rho = \gamma \sum_{j} \left( \eta^{+}_{j} \rho \eta^{-}_{j} - \frac{1}{2} \left\{\eta^{-}_{j}\eta^{+}_{j}, \rho\right\} \right)
$$

which has the exactly the same form as the original one.  We will be using this transformation in some applications.

Although the expression changed, the density operator is invariant. So, if we are interested in only the state of the emitter, there seems no problem.  The expectation value of any observabale in the emitter's Hilbert space is indeed invariant. 

However, we are interested in the light (photon) emitted by the emitters. The light is not included in the master equation.  During the derivation of the master equation, we thrown away the information of the radiation field.   Actually the loss of information on the radiation field is the caused of ambiguity in the quantum master equation.   The detector obviously interacts with the radiation field, not with the emitters.  Therefore, the solution of the quantum master equation has no direct connection to the detector.  What is the information missing in the solution to the master equation?  First of all, it does not tell which transition emits a photon.  The concept of the transition depends on the basis set.  In turn, we don't know what mode of photon is emitted.  At the end of the process, the detector picks up a photon in a certain mode, which depends of the experimental set up. None of those information is included in the quantum master equation.  The ambiguity is precisely due to the lack of these information.

In the jump method which will be introduced in {numref}`ch:openqm-jump`, we will learn that the detector sees a jump operator (collapse operator).  As shown in {numref}`ch:openqm-jump`  the jump operator used by the detector is the correct Lindblad operator in the  master equation.  In turn the jump operator that the detector sees is determined by the radiation field emitted by the emitter like Eq. {eq}`eq:source_field_approx`.  At the end, we must pick a correct jump operator consistent with physics, in particular quantum entanglement between the emitter and the radiation field.


## Detection of photons

In {numref}`ch:detector`, we learned the Glauber's theory of detection.  A history of clicks recorded by a detector corresponds to the measurement of radiation intensity.  However, a quantum master equation does not have that information.  Folloiwng Lehmberg {cite:p}`Lehmberg1970a`,  we can make a connection between the density matrix and intensity through the source radiation field  {eq}`eq:source_field_approx`.    The intensity at the position of detector $\mathbf{R}$ is given by

$$
I(\mathbf{x},t) = \langle E^{(-)}(\mathbf{x},t)E^{(+)}(\mathbf{x},t)\rangle \rangle =  
\langle J^{+}(\mathbf{R},t) J^{-}(\mathbf{R},t) \rangle
$$


where

$$
J^{\pm} (\mathbf{R},t) = \sqrt{\gamma(\mathbf{R})} \sum_{i} e^{ \pm i k_{0} \hat{\mathbf{R}} \cdot \mathbf{r}_{j}} \sigma^{\pm}_{j}(t).
$$(eq:jump-operator-R)

with the transition rate 

$$
\gamma(\mathbf{R}) = \left(\frac{\omega_{0}^{2}d}{c^2}\right)^{2} \left [ 1 - \left(\hat{\mathbf{d}}\cdot\hat{\mathbf{R}}\right)^2\right].
$$

This suggest that $J^{-}$ is the jump operator the detector measures (in the Heisenberg picture) if the all the emitters are coherent. 

If the emitters are not coherent,

$$
\begin{aligned}
I(\mathbf{x},t) &= \langle J^{+}(\mathbf{R},t) J^{-}(\mathbf{R},t) \rangle \\
&= \left(\frac{\omega_{0}^{2}d}{c^2}\right)^2 \left [ 1 - \left(\hat{\mathbf{d}}\cdot\hat{\mathbf{R}}\right)^2\right] \sum_{ij} e^{i k_0 \hat{\mathbf{R}}\cdot \mathbf{r}_{ij}} \langle \sigma^{+}_{i} \sigma^{-}_{j}\rangle \\
&= \left(\frac{\omega_{0}^{2}d}{c^2}\right)^2 \left [ 1 - \left(\hat{\mathbf{d}}\cdot\hat{\mathbf{R}}\right)^2\right] \sum_{j}  \langle \sigma^{+}_{j} \sigma^{-}_{j}\rangle 
\end{aligned}
$$

where we used the incoherent condition $\langle \sigma^{+}_{i} \sigma_{j}\rangle = 0$ for $i\ne j$.  In this case, $\sqrt{\gamma(\mathbf{R})} \sigma^{-}_{j}$ is the jump operator.  To convert this jump operator for the quantum master equation, we need to count all photons emitted in all direction.  Integrating $\gamma(\mathbf{R})$ over the solid angle

$$
\int_{0}^{2\pi} \int_{-\pi}^{\pi} \gamma(\mathbf{R}) \sin(\theta) d\theta d\phi = \gamma_0
$$

which is the standard rate of spontaneous emission.

## Coherence function

We now measure the second-order coherence function by a detector placed at $\mathbf{R}$.  Using the jump operators {eq}`eq:jump-operator-R`, 

$$
g^{(2)}(\tau;\mathbf{R}) 
= \frac{\langle J^{+}(\mathbf{R},t) J^{+}(\mathbf{R},t+\tau) J^{-}(\mathbf{R},t+\tau) J^{-}(\mathbf{R},t)  \rangle_{ss}}{(\langle J^{+}(\mathbf{R},t) J^{-}(\mathbf{R},t)\rangle_{ss})^2} =
= \frac{\sum_{ij} \sum_{k\ell} e^{i k_0 \hat{\mathbf{R}}\cdot \mathbf{r}_{ij}}e^{i k_0 \hat{\mathbf{R}}\cdot \mathbf{r}_{k\ell}} \langle\sigma^{+}_{i}(t) \sigma^{+}_{k}(t+\tau)\sigma^{-}_{\ell}(t+\tau)\sigma^{-}_{j}(t)\rangle_{ss}}{ \left( \sum_{ij} e^{i k_0 \hat{\mathbf{R}}\cdot \mathbf{r}_{ij}}\langle\sigma^{+}_{i}(t)\sigma^{-}_{j}(t)\rangle_{ss}\right)^2 }.
$$(eq:g2-src-field)

which assume that the emitters are all coherent.  if not we use  $\sqrt{\gamma} a^{-}_{j}$ as jump operators and we obtain

$$
g^{(2)}(\tau;\mathbf{R}) 
= \frac{\sum_{ij}  \langle\sigma^{+}_{i}(t) \sigma^{+}_{j}(t+\tau)\sigma^{-}_{j}(t+\tau)\sigma^{-}_{i}(t)\rangle_{ss}}{ \left( \sum_{j} \langle\sigma^{+}_{j}(t)\sigma^{-}_{j}(t)\rangle_{ss}\right)^2 }.
$$(eq:g2-src-field-incoherent)
