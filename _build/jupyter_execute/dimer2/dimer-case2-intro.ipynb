# Dimer II: Coherence by indistinguishability

In {numref}`ch:coherence`, we discussed the origin of decoherence.  When the radiation field can't distinguish the emitters, coherence survives when the partial trace is taken. The field very close to the emitters (near field) can obviously see them.   So, when the radiation field failed to see individual atoms.  The field far from the emitters (far field) is expressed by Eq.{eq}`eq:source_field_approx`.  The difference between the emitters appears in the phase factor, 
$e^{- i k_{0} \hat{\mathbf{R}} \cdot (\mathbf{r}_{1}-\mathbf{r}_{2})}$.   Roughly speaking, if $r_{12} k_{0} \ll 1$, then the phase difference vanishes and the far field can't distinguish the emitters.

In this section, we assume that the field can't distinguish the emitters.  Then, a dark state introduced in {numref}`ch:coherence` need to be taken into account, which results in _superradiace_ and _subradiance_.
