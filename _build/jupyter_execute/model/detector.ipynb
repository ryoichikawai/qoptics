# Dectector

It is important to understand what physical quantity the detector is measuring. In the standard quantum mechanics, a physical quantity is represented by a self-adjoint operator.  The outcome of the measurement is one of the eigenvalues of the operator. At the end of the measurement, the system state collapses (is projected) to the corresponding eigenvector. The probability to obtain a particular outcome is determined by the Born rule.  This kind of measurement is known as [projective measurement](https://en.wikipedia.org/wiki/Measurement_in_quantum_mechanics).

Measurement of radiation fields or photons are totally different from the projective measurement. First of all, the photon disappears when it was detected.  Therefore, the projection is not possible.  Secondly, photons do not have position.  Mathematically we cannot define the position operator for photons. 
At the foundamental level, the quantum measurement in the quantum field theory is still debated. We will not discuss it here.  

For most practical applications,  the photon detection model developed by Glauber {cite:p}`Glauber2007` works quite well.   Briefly stating, a detector can measure only the intensity of radiation field through a record of clicks.  Nothing else.  See also Carmichael {cite:p}`Carmichael1993` for the discussion of photodetection.

## The Glauber model

First of all, we need to think what a photodetector records.  It clicks when a photon is detected.  The outcome of measurement is not a usual physical quantity.   The outcome is "0" for no signal and "1" for a click.  A single click does not have any useful information except that there was at least one photon in the space.   After many clicks, we have a random time series.  What we can learn from it?  The theory of statistics can extract many information out of it. (See {numref}`ch:bunching-fcs`.) We are particularly interested in two of them. 1) how often the clicks happen (mean detection rate) and 2) stndard deviation form the mean.

At each click, an electron in the detector absorb a photon.  Therefore, the mean detection rate corresponds to the absorption rate which we can calculate using the standard quantum mechanics.  Usually we use a perturbation theory (the Fermi's golden rule). Glauber was able to connect the transition rate to the mean intensity of radiation. {cite:p}`Glauber2007`   We now show that the detector measures the mean intensity {eq}`eq:radiation-intensity`.

## Transion rates

Suppose that the radiation field is in a state $|i\rangle$ and the electron in $|g\rangle$ before the click and $|f\rangle$ and $|e\rangle$ after the click.  The transition rate is proportional to the square of the transition matrix $W_{if} \propto |\langle e |\otimes \langle f | V | |g\rangle \otimes | i \rangle|^2$ where $V$ is the interaction Hamiltonian defined as (See Eq. {eq}`eq:SR-coupling`.)

$$
V = - \left (\mathbf{E}^{(+)}(\mathbf{x}) + \mathbf{E}^{(-)}(\mathbf{x})\right) \cdot \mathbf{d} \left (\sigma^{-} + \sigma^{+}\right)
$$

where $\mathbf{x}$ is the location of the detector.

Evaluating the matrix element, we obtain

$$
W_{if} \propto \left | \langle f | \mathbf{E}^{(+)}(\mathbf{x}) | i \rangle \right|^2
$$

Since we don't know what is the final state $f$, we consider all possible state (the initial state automatically picks up the correct one due to the orthogonality).  Summing up over all $|f\rangle$, the detection rate from $|i\rangle$ is

$$
W_{i} \propto  \sum_{f} \left | \langle f | \mathbf{E}^{(+)}(\mathbf{x}) | i \rangle \right|^2
=   \sum_{f} \langle i | \mathbf{E}^{(-)}(\mathbf{x}) | f \rangle   \langle f | \mathbf{E}^{(+)}(\mathbf{x}) | i \rangle 
= \langle i | \mathbf{E}^{(-)}(\mathbf{x}) \mathbf{E}^{(+)}(\mathbf{x}) | i \rangle
$$

Finally, we average over all possible initial state with probability $P_{i}$, the final form of the transition rate

$$
W \propto \sum_{i} P_{i} \langle i | \mathbf{E}^{(-)}(\mathbf{x}) \mathbf{E}^{(+)}(\mathbf{x}) | i \rangle = \text{tr} \left[ \mathbf{E}^{(-)}(\mathbf{x}) \mathbf{E}^{(+)}(\mathbf{x}) \rho \right ] = \langle \mathbf{E}^{(-)}(\mathbf{x}) \mathbf{E}^{(+)}(\mathbf{x}) \rangle.
$$

The transition rate (detection rate) is proportional to the intensity of the radiation. We conclude that detector is essentially measuring the radiation intensity.

## Connection to the emitter

When a photon is detected by the detector, the state of the emitter must collapse to a corresponding state.  From the clicking data only, we don't know what happened to the emitter state.  Theoretically we can unravel the corresponding collapse.  That is one of the goal of this lecture.
