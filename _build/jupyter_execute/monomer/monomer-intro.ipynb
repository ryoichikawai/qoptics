# Monomer: Driving and the Rabi oscillation

In {numref}`sec:openqm-qme-example` and {numref}`sec:openqm-jump-example`, we investigated the thermal fluorescence using the open quantum system approach. In this chapter, we investigate a single emitter driven by an time-dependent external field (See {numref}`ch:driving` for the model of driving.) The system Hamiltonian is no longer diagonal in the energy basis due to the driving and thus the situation is slightly different from the example discussed in {numref}`sec:openqm-qme-example` and {numref}`sec:openqm-jump-example`. Since the driving is stronger than thermal excitation, we can neglect thermal effect.  First we introduce a mathematical model.  Then, the time-evolution of the system and second order coherence are investigated by two different approaches, 1) quantum master equation, and 2) quantum trajectory. 

The main consequence of the driving is the appearance of the Rabi oscillation in the system evolution and in the first/second-order coherence.  Hence, the focus on this chapter is the Rabi oscillation.
