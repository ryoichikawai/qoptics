# The second-order coeherence function

The first-order coherence function quantifies the coherence of amplitude, in direct connection to interference fringes.  The experiment by Hanbury Brown and Twiss demonstrated that a different type of coherence exists. It is a spatio-temporal correlation of intensity. 

## Definition

The second-order coherence function is defined as correlation of radiation intensity at two different spacetime points:

$$
\begin{aligned}
g^{(2)}(t_{1},t_{2};\mathbf{x}_{1},\mathbf{x}_{2}) 
&= \frac{\langle I(\mathbf{x}_{1},t_{1}) I(\mathbf{x}_{2},t_{2})\rangle}{\langle I(\mathbf{x}_{1},t_{1}) \rangle\langle I(\mathbf{x}_{2},t_{2}) \rangle}\\
&= \frac{\langle\big(\mathbf{E}^{(-)}(\mathbf{x}_{1},t_{1}) \cdot \mathbf{E}^{(+)}(\mathbf{x}_{1},t_{1})\big)\big(\mathbf{E}^{(-)}(\mathbf{x}_{2},t_{2}) \cdot \mathbf{E}^{(+)}(\mathbf{x}_{2},t_{2})\big)\rangle}{\langle\mathbf{E}^{(-)}(\mathbf{x}_{1},t_{1}) \cdot \mathbf{E}^{(+)}(\mathbf{x}_{1},t_{1})\rangle \langle\mathbf{E}^{(-)}(\mathbf{x}_{2},t_{2}) \cdot \mathbf{E}^{(+)}(\mathbf{x}_{2},t_{2})\rangle}
\end{aligned}
$$(eq:g2-def)

In many case, we are interested in the temporal correlation measured at a single point

$$
\begin{aligned}
g^{(2)}(\tau;\mathbf{x})
&= \frac{\langle I(\mathbf{x},t) I(\mathbf{x},t+\tau)\rangle}{\langle I(\mathbf{x},t) \rangle\langle I(\mathbf{x},t+\tau) \rangle }\\
&= \frac{\langle I(\mathbf{x},t) I(\mathbf{x},t+\tau)\rangle}{[\langle I(\mathbf{x}) \rangle]^2}\\
\end{aligned}
$$(eq:g2-tau-x1)

In the last line, we assume that the system is in a steady state.  We will mainly work on this temporal correlation.

## Some general properties

We expect that the correlation vanishes for a large $\tau$, that meansfor large $\tau$,  $\langle I(\mathbf{x},t) I(\mathbf{x},t+\tau)\rangle \approx \langle I(\mathbf{x},t) \rangle \langle I(\mathbf{x},t+\tau) \rangle$, from which we obtain 

$$
\lim_{\tau\rightarrow\infty} g^{(2)}(\tau;\mathbf{x}) = 1.
$$

When $\tau=0$ (nodelay correlation function), 

$$
g^{(2)}(0;\mathbf{x}) = \frac{\langle I^{2}(\mathbf{x},t)\rangle}{\langle I(\mathbf{x},t) \rangle^2} = 1 + \left(\frac{\Delta I(\mathbf{x},t)}{\langle I(\mathbf{x},t) \rangle}\right)^{2}
$$(eq:g2-nodelay-classical)

where the variance is defined by $\left(\Delta I(\mathbf{x,t})\right)^{2} = \langle I^{2}(\mathbf{x},t) \rangle - \langle I(\mathbf{x},t)\rangle^{2} \ge 0$.  Equation {eq}`eq:g2-nodelay-classical` indicates that $g^{(2)}(0,\mathbf{x}) \ge 1$.  The equality holds if and only if $\Delta I =0$.  This means that if an experiment data shows $g^{(2)}(0)=1$ then the light intensity is not fluctuating. The positive value was first confirmed by Hanbury Brown and Twiss.{cite:p}`HanburyBrown1956`.  More recently, cleaner results have been obtained  with  modern experimental techniques.{cite:p}`Tan2014,Tan2017`.

It turns out that negative value of $g^{(2)}(0)$ has also been observed{cite:p}`Kimble1977,Grangier1986` in violation of the above result.  Such results can be explained only by quantum optics.


## Quantum effects

We assume that the system is in a steady state and thus $g^{(2)}$ depends only the time differece $\tau = t_{2}-t_{1}$.  Using Eqs. {eq}`eq:quantum-free-field-pm` and {eq}`eq:radiation-intensity`, Eq. {eq}`eq:g2-def` can be expressed as

$$
g^{(2)}(\tau;\mathbf{x}_1,\mathbf{x}_2) = \sum_{ij} \sum_{i'j'} w_{ij}(\mathbf{x}_1) w_{i'j'}(\mathbf{x}_2) g^{(2)}_{ij,i'j'}(\tau)
$$(eq:g2-quantum)

with

$$
g_{ij,i'j'}^{(2)}(\tau) = \frac{\langle a^{\dagger}_{i'}(t) a^{\dagger}_{i}(t+\tau) a_{j}(t+\tau) a_{j'}(t)\rangle_{ss}}{\langle a^{\dagger}_{i}(t) a_{j}(t)\rangle_{ss} \langle a^{\dagger}_{i'}(t) a_{j'}(t)\rangle_{ss}}
$$(eq:subg2-quantum)

See {numref}`sec:coherence-g1` for $w_{ij}$.  

Pay a close attention to the order of creation and annihilation operators, which is a crucial point of quantum correlation.  In the product of creation and annihilation operators, all creation operators are placed to the left of all annihilation operators to eliminate infinite zero-point vacuum energy and resolve ordering ambiguities when promoting classical fields to quantum operators. This order is called [normal order](https://en.wikipedia.org/wiki/Normal_order).  The rule of the normal order does not tell how to order chronologically.  That is determined by the detector model by Glauber. (See {numref}`ch:detector`.) Recall that each click corresponds to a jump expressed by an annihilation operator.  If the first click happened at $t_1$ and the second one at $t_{2}$, the state jumps as $a(t_{2}) a(t_{1})$ (the right most is the earliest).  To compute the transition probability we need its adjoint $[a(t_2) a(t_1)]^{\dagger} = a^{\dagger}(t_{1}) a^{\dagger}(t_{2})$.  Combining all, we have $a^{\dagger}(t_{1}) a^{\dagger}(t_{2})a(t_{2}) a(t_{1})$.  REquirement from the quantum field theory and the detection theory uniquely determine the order of operators.

Let us look at the simplest case, no-delay coherence of a single mode radiation:

$$
g^{(2)}(0;\mathbf{x}) = \frac{\langle a^{\dagger} a^{\dagger} a a\rangle_{ss}}{(\langle a^{\dagger}a \rangle_{ss})^2}
$$(eq:g2-quantum-single-mode)

which does not depend on the location. $\langle \cdots \rangle_{ss}$ indicates average over a steady state.  Using the commutation relation, $a^{\dagger}a^{\dagger} a a = a^{\dagger} a a^{\dagger} a - a^{\dagger} a = n^2-n$, we find

$$
\begin{aligned}
g^{(2)}(0) &= \frac{\langle n(n-1) \rangle_{ss}}{\langle n \rangle_{ss}^2} = \frac{\langle n^2 \rangle_{ss}}{\langle n \rangle_{ss}^2} - \frac{1}{\langle n \rangle_{ss}} \\
&=1 + \left(\frac{\Delta n}{\langle n \rangle_{ss}}\right)^2 - \frac{1}{\langle n \rangle_{ss}}
\end{aligned}
$$(eq:nodelay-g2-quantum-single-mode)

where we used the fluctuation of photon numbers $(\Delta n)^2 = \langle n^2 \rangle_{ss} - \langle n \rangle_{ss}^2$. This nodelay second-order coherence function looks similar to the classical version {eq}`eq:g2-nodelay-classical` but actually quite different.  While the classical value is always above 1, we cannot say so for the quantum version.  For example, if $\langle n \rangle_{ss}=1$ and $\Delta n = 0$, then we have $g^{(2)}(0) = 0$.  That means if the detector sees only one photon at a time, the nodelay second-order coherence function vanishes.  This is purely a quantum effect known as photon antibunching.

If the order of the operators are different, we did not get this result. It is critical to get the correct order of creation annihilation operators to obtain correct quantum effects.

```{admonition} Order of operators
:class: important
:name: adm:normal-oirder

The creation and annhilation operators must be ordered as follows.

1. **Normal order**: All creation operators must be placed to the left of all annihilation operators.
2. **Chronological order**: Annihilation operators must be ordered chronologicaylly according to the clicks of the dector.
```

## Is $g^{(2)}$ measurebale?

Unlike $g^{(1)}$, there is no difficulty in measuring $g^{(2)}$ since intensity is an observable and recording of clicks by the detector can be converted to the intensity.  See {numref}`ch:detector`.
