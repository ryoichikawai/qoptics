# First-order coherence function

## Definition

The first-order coherence function is defined as correlation of the field amplitude at two different spacetime points: 

$$
g^{(1)}(t_{1},t_{2};\mathbf{x}_1,\mathbf{x}_2) = 
\frac{\langle\mathbf{E}^{(-)}(\mathbf{x}_{1},t_{1}) \cdot \mathbf{E}^{(+)}(\mathbf{x}_2,t_{2})\rangle}
{\sqrt{\langle\mathbf{E}^{(-)}(\mathbf{x}_{1},t_{1})\cdot\mathbf{E}^{(+)}(\mathbf{x}_{1},t_{1})\rangle \langle\mathbf{E}^{(-)}(\mathbf{x}_{2},t_{2})\cdot\mathbf{E}^{(+)}(\mathbf{x}_{2},t_{2})\rangle}}
$$(eq:g1-full)

where $(\mathbf{x}_1,t_{1})$ and $(\mathbf{x}_2,t_{2})$ are two "measurement" points.  This function is complex. The modulus $\gamma^{(1)} \equiv |g^{(1)}|$, known as the degree of first coherence, is bound between $0$ and $1$. A value of $1$ indicates perfect coherence, while $0$ means the fields are completely uncorrelated.

We will focus on its special case, temporal correlation measured at a single location:

$$
g^{(1)}(\tau;\mathbf{x}) = 
\frac{\langle\mathbf{E}^{(-)}(\mathbf{x},t) \cdot \mathbf{E}^{(+)}(\mathbf{x},t+\tau)\rangle}
{\langle\mathbf{E}^{(-)}(\mathbf{x},t)\cdot\mathbf{E}^{(+)}(\mathbf{x},t)\rangle}.
$$(eq:g1-temp)

which measures _coherence time_ (how long the radiation field remembers its phase.)

$$
\lim_{\tau\rightarrow\infty} g^{(1)}(\tau;\mathbf{x}) = 0
$$

and 

$$
g^{(1)}(0;\mathbf{x}) = 1
$$

at any position.

## Quantum expression

In quantum oprics, it is more convenient to write Eq. {eq}`eq:g1-temp` with photon operators. Substituting {eq}`eq:quantum-free-field-pm`, Eq. {eq}`eq:g1-temp` is expressed as

$$
g^{(1)}(\tau;\mathbf{x}) = \frac{\sum_{i,j} f_{ij}(\mathbf{x}) \langle a^{\dagger}_{i}(t+\tau) a_{j}(t)\rangle}{\sum_{i,j} f_{ij}(\mathbf{x})  \langle a^{\dagger}_{i}(t) a_{j}(t)\rangle}
$$(eq:g1-quantum)

where

$$
f_{ij}(\mathbf{x}) = \mathbf{e}_{i}\cdot \mathbf{e}_{j}\, \mathcal{E}_{i} \mathcal{E}^{*}_{j} \, e^{i (\mathbf{k}_{i}-\mathbf{k}_{j})\cdot \mathbf{x}}.
$$

To interpret Eq. {eq}`eq:g1-quantum` we rewrite it in slightly different expression:

$$
g^{(1)}(\tau,\mathbf{x}) = \sum_{ij} w_{ij}(\mathbf{x}) g^{(1)}_{ij}(\tau).
$$(eq:g1-weighted-sum)

where 

$$
g^{(1)}_{ij}(\tau) = \frac{\langle a^{\dagger}_{i}(t+\tau)a_{j}(t)\rangle_t}{\langle a_{i}^{*}(t)a_{j}(t)\rangle}
$$(eq:g1-temp-only)

and weight functions

$$
w_{ij}(\mathbf{x}) = \frac{f_{ij}(\mathbf{x},\mathbf{x})  \langle a^{\dagger}_{i}(t) a_{j}(t)\rangle}
{\sum_{ij}f_{ij}(\mathbf{x},\mathbf{x})  \langle a^{\dagger}_{i}(t) a_{j}(t)}.
$$(eq:wij)

The weight is normalized as $\sum_{ij} w_{ij}(\mathbf{x}) = 1$.  Equation {eq}`eq:g1-weighted-sum` indicates that at each oint $\mathbf{x}$, the contribution of temporal correlation functions $g^{(1)}_{ij}$ is different.

## Is $g^{(1)}$ measurable?

In classical world, the temporal correlation $\langle A(t+\tau) B(t) \rangle$ can be measured directly.  We record the time evolution of $A(t)$ and $B(t)$. Then, we make statistical analysis of the records.  In quantum mechanics, that approach is not possible because the measurement changes the state.  The situation is even worse for $\langle a^{\dagger}(t+\tau) a(t) \rangle$ because $a(t)$ is not an observable.  Furthermore, the photodetector can measure only the intensity of the field.  Then, how can we determine $g^{(1)}(\tau)$ by experimental measurement?

There is a method using an interferometer.  The actual procedure is similar to [quantum tomography](https://en.wikipedia.org/wiki/Quantum_tomography).  Suppose that a beam of light is divided into two beams, A and B, by a beam splitter. Beam B travles a longer path than A.  At time $t$ the radiation fields of beam A and B at the detector are $\mathbf{E}^{\pm}_{A}(t) = \frac{1}{\sqrt{2}} = \mathbf{E}^{\pm}(t)$ and $\mathbf{E}^{\pm}_{B}(t) = \frac{1}{\sqrt{2}} = \mathbf{E}^{\pm}(t+\tau)$ where $\tau$ is the delay time due to the longer path.   The detector measures the intensity of the combined beam, $\mathbf{E}^{\pm}_{D}(t) = \frac{1}{\sqrt{2}} \left[\mathbf{E}^{\pm}(t) + \mathbf{E}^{\pm}(t+\tau) \right]$.  Then, the intensity observed by the detector is

$$
\begin{aligned}
I_{D}(\tau) &= \langle \mathbf{E}^{(+)}_{D}(t)\cdot\mathbf{E}^{(-)}_{D}(t)\rangle \\
& = \frac{1}{2} \langle \left[\mathbf{E}^{+}(t) + \mathbf{E}^{(+)}(t+\tau)\right]\cdot 
\left[\mathbf{E}^{-}(t) + \mathbf{E}^{(-)}(t+\tau)\right]\rangle \\
& = I_{A} + I_{B}
+ \text{Re} \langle \mathbf{E}^{(+)}(t)\cdot\mathbf{E}^{(-)}(t+\tau)\rangle \\
& = I_{A} + I_{B} + \text{Re}\, G^{(1)}(\tau) 
\end{aligned}
$$

where $I_{D}$, $I_{A}$, and $I_{B}$ are the intensity at the detector and the intensities of beam A and B.  From the conservation of energy $I_{A}+ I_{B}=I_{0}$ where $I_{0}$ is the intensity of the original beam before splitting.  Then,

$$
\begin{aligned}
\frac{I_{D}(\tau)}{I_{0}} &=  1 + \text{Re}\, g^{(1)}(\tau) \\
& = 1 + \gamma^{(1)}(\tau) \cos(\phi(\tau))
\end{aligned}
$$

where $\gamma^{(1)}(\tau)$ and $\phi(\tau)$ are modulus and phase of $g^{(1)}(\tau)$.  In a regular interference, the fringes are generated by the phase and the modulus does not change.  Therefore, we assume that $\gamma^{(1)}$ is constant.

To determine the real part of $g^{2}$, we need to measure $I_{0}$ and $I_{D}$.  We need to measure them separately in the same experimental setup since the measurement of $I_{0}$ change the outcome of $I_{D}$.

To find the imaginary part, we change $\tau$ (change the traveling distance of beam B) and find the maximum and minimum intensity

$$
\begin{aligned}
\frac{I^\text{max}_{D}}{I} &= 1 + \gamma^{(1)} \\
\frac{I^\text{min}_{D}}{I} &= 1 - \gamma^{(1)}
\end{aligned}
$$

From which we can find the modulus (known as degree of first-order coherence),

$$
\gamma^{(1)} = \frac{I^\text{max}_{D} - I^\text{min}_{D}}{2 I_0}
$$

Now it is straight forward to compute the phase.

In conclusion, it is possible to determine $g^{(1)}(\tau)$ by a single photon detector.  However, it is not a simple measurement.  We need to measure intensity  by changing the travel distance.  In other words, we need to map interference fringes by changing parameters.  We need also to be careful not to destroy the interference fringes.

For more detailed discussion, see Ref. {cite:p}`Ficek2005`.
