# What is quantum coherence?

## Definition of coherence

In {numref}`ch:photons`, we mentioned the difficulty of defining the concept of photon.  Similarly, the precise definition of quantum coherence is difficult.  Most of textbooks do not touch on this issue. People seems agreeing that Coherence is the ability of a state to generate observable interference.


```{prf:definition} Coherence
:label: def:coherence
Coherence is the ability of a state to generate observable interference.
```

How do we mathematically implement the above definition?  The state of a quantum system is expressed as a density operator $\rho$.  How do we know if a density operator is capable of generating observable interference?  We need to make a certain measurement to see interference.
Suppose that we measure an observable $\hat{A}$ whose eigenvalues and the corresponding eigenvectors are expressed as $a_{j}$ and $|a_{j}\rangle$.  Based on the Born rule, the probability of obtaining $a_{1}$ is given by 

$$
P_{1} = \text{tr}\left(|a_{1}\rangle\langle a_{1}| \rho\right) = \langle a_{1}|\rho|a_{1}\rangle
$$

This expression does not expose the presence of coherence.  We need to introduce a basis set $|u_{j}\rangle$.  Then, the probability is written as

$$
P_{1} = \sum_{ij} \langle a_{1} | u_{i} \rangle \rho_{ij} \langle u_{j} | a_{1}\rangle = \sum_{j} |\langle a_{1}|u_{j}\rangle|^2 \rho_{jj} + \sum_{i\ne j} \langle a_{1} | u_{i} \rangle \langle u_{j} | a_{1}\rangle \rho_{ij}
$$

where the density matrix is defined by $\rho_{ij} = \langle u_{i} | \rho | u_{j}\rangle$.  The second term in the right hand side is clearly showing interference between $|u_{1}\rangle$ and $|u_{2}\rangle$ through the off-diagonal elements $\rho_{i\ne j}$.  Therefore, mathematically speaking, if the state has the ability to generate obervable interference or not is determined by the off-diagonal elements of the density operator.  

But it depends on the choice of the basis set. If a different basis set is used, the value of the off-diagonal elements change.  Furthermore, we can always diagonalize a density matrix, eliminating all off-diagonal elements entirely.  

We conclude that  coherence is the ability of a state to generate observable interference between _certain basis vectors_.  The ability is determined by the off-diagonal elements of the density matrix in _a specified basis set_.

```{admonition}  Coherence and basis set
:class: important
:name: adm:coherence-basis

It does not make a sense to discuss coherence without specifying a basis set. 
```

## Coherence vs superposition

Interference is well-defined phenomena in the classical wave theory. In quantum mechanics, wave functions can generate interference exactly in the same way as the classical waves.  It seems that any quantum state can generate interference and thus any quantum state seems coherent.  This misleading idea is coming from undergraduate quantum mechanics courses in which all examples are perfectly coherent.

In undergraduate textbooks, the state of a quantum system is mathematically specified by a vector in a Hilbert space.  The superposition principle states that any state vector $|\psi\rangle$ can be expressed as a superposition of basis vectors $|u_{j}\rangle$ as

$$
|\psi\rangle = \sum_{j} c_{j} |u_{j}\rangle
$$

Let us check if this state is coherent or not based on our definition.  Writing the state in terms of density operator

$$
\rho = |\psi\rangle \langle \psi | = \sum_{ij} c_{i} c^{*}_{j} |u_{i}\rangle \langle u_{j}|
$$

The off-diagonal elements are $\rho_{i\ne j} = c_{i} c^{*}_{j}$. All off-diagonal elements vanish if and only if $c_{i}\ne 0$ and all other coefficients vanish, a state $|\psi\rangle = |u_{i}\rangle$ cannot generate interference in the current basis set.[^1]  Otherwise, the state has the ability of generating interference and thus the state is coherent.   Does this mean any quantum state is coherent?

It is true that if the state of a system can be written as a superposition, then the system is coherent in that basis.  No doubt about that.  The question is if all states can be written as a superpoisition of basis vectors.  The answer is No.  Any state that can be written as a state vector is called a _pure_ state and the corresponding density operator is $\rho = |\psi\rangle|\langle \psi |$.  There are other states that cannot be expressed with a state vector.  Those states are known as  a _mixed_ state.  A mixed state can be also coherent even though  it cannot be in a simple superposition state.

In conclusion, any pure state written as a superposition of basis vectors, it is highly coherent. However, not all states are pure. A mixed state can have a certain level of coherency  as long as not all off-diagonal elements vanish.

[^1]: If there is only one nonvanishing diagonal element, the state is pure and we have no way to determine coherency from it.

## 

The origin of decoherence

The fundamental principles of quantum mechanics defines the state of a system as a vector which is inherently coherent in any basis.  Then, how is the coherence lost?  Actually, we never lose coherence if the whole system including environments are included in the quantum description.

TO see that, consider a pair of atoms, $A_{1}$ and $A_{2}$ each has a ground state $|g\rangle$ and an excited state $|e\rangle$.  $A_{1}$ emits radiation $|R_{1}\rangle$ and independently $A_{2}$ emits $|R_{2}\rangle$.  Suppose that after one photon is emitted, the total system is in $|\Psi\rangle = \frac{1}{\sqrt{2}} \left(|ge\rangle \otimes |R_{1}\rangle + |eg\rangle \otimes |R_{2}\rangle\right)$, where $|ge\rangle$ indicates $A_{1}$ in the ground state and $A_{2}$ in the excited state. $|\Psi\rangle$ is highly entangled. Now, we check  if the interference fringes is formed.

$$
\begin{aligned}
|\Psi(x)|^{2} &= \text{tr}\left[(I \otimes | x \rangle\langle x |) |\Psi\rangle\langle\Psi|\right] \\
& = \frac{1}{2} \left(|R_{1}(x)|^2 +|R_{2}(x)|^2 
+ \langle ge|eg\rangle R^{*}_{1}(x) R_{2}(x) + \langle eg|ge\rangle R^{*}_{2}(x) R_{1}(x)\right)\\
&= \frac{1}{2} \left( |R_{1}(x)|^2 + |R_{2}(x)|^2 \right)
\end{aligned}
$$

where $R_{j}(x) = \langle x | R_{j}\rangle$.  The orthogonality between $|ge\rangle$ and $|eg\rangle$ destroyed the interference terms. The fact that each radiation field $R_{j}$ is tightly correlated to the corresponding atom $A_{j}$ is the cause of decoherence.   In quantum information theory, we say  that the emission of the radiation is recorded in each atom.

We consider another state $|\Phi\rangle = \frac{1}{\sqrt{2}}(|eg\rangle + |ge\rangle) \otimes \frac{1}{\sqrt{2}}(|R_{1}\rangle + |R_{2}\rangle )$ which has no entanglement between the atoms and the radiations.  No information about which atom emitted which radiation is recorded. The probability distribution for this state is 

$$
\begin{aligned}
|\Phi(x)|^2 &= \text{tr}\left[(I \otimes | x \rangle\langle x |) |\Phi\rangle\langle\Phi|\right] \\
& = \frac{1}{2} |R_{1}(x)+R_{2}(x)|^2 \\
& = \frac{1}{2} \left[ |R_{1}(x)|^2 + |R_{2}(x)|^2
+ R^{*}_{1}(x) R_{2}(x) + R_{1}(x) R^{*}_{2}(x) \right]
\end{aligned}
$$

which shows usual interference fringes.

These two extreme examples show that even when the whole system is in a pure state (coherent), the interference can vanish depending on the structure of state vectors.   If the state is in a product state, perfect interference is observed.  On the other hand, if the state is entangled, interference is partially or entirely destroyed depending on the type of entanglement.  

For detailed discussion on decoherence beyond optics, see Refs. {cite:p}`Breuer2002,Schlosshauser2007,Zurek2025`.


```{admonition} Decoherence
:class: important
:name: adm:decoherence

The coherence is lost when the system is quantum mechanically entangled with other systems or environments. 
```

## Subsystems and reduced density operator

At the beginning of this page, we showed that the off-diagonal elements of the density operator are an good indicator of coherence.  Then, we showed that entanglement between the system and the radiation field is the origin of decoherence.  How do these two conclusions can be reconciled?

The problem is that the first picture focuses only on the atoms whereas the second view involves both the atoms and the radiation fields.   In other words, the first view is looking at a subsystem.  In the classical systems, the whole system is just a sum of subsystems.  In quantum mechanic, you cannot express a state of the whole system by adding or multiplying the states of subsystems.

Mathematically speaking a state of a subsystem is defined by a partial trace.  Consider a bipartite system consisting of A and B.  The state of the total system is given by a density matrix $\rho_{AB}$.  Then, what is the state of subsystem A?  It is defined by
a reduced density operator defined by a partial trance {cite:p}`Audretsch2007`:

$$
\rho_{A} = \text{tr}_{B} \rho_{AB} = \sum_{j} \left(I_{A} \otimes \langle u_{j} | \right) \rho_{AB} \left(I_{A} \otimes | u_{j} \rangle \right)
$$

where $I_{A}$ is the identity operator in the Hilbert space for A and $|u_{j}\rangle$ are orthonormal basis kets in the Hilbert space for B.  Any basis set can be used.  Note that the partial trace $\text{tr}_{B}$  hides the information stored in the subsystem B.  Thus, we can not get any information on B by lokking at $\rho_{A}$.

Now, we find the reduced density for atoms by tracing over the radiation fields.

For the entangled case $|\Psi\rangle$, the reduced density for the atoms 

$$
\rho_{A} = \text{tr}_{R} |\Psi\rangle\langle\Psi| = |ge\rangle\langle ge| + |eg\rangle\langle eg| + |ge\rangle\langle eg| \langle R_{2} | R_{1}\rangle +
|ge\rangle\langle ge| \langle R_{1} | R_{2}\rangle
$$

The off-diagonal element is $\rho_{eg,ge} = \langle R_{1} | R_{2}\rangle < 1$.  If $|R_{1}\rangle $ and $|R_{2}\rangle$ are orthogonal, it completely vanish. Hence, the coherency is entirely lost.  Other wise, the coherency is only partial.

On the other hand, for the product state $|\Phi\rangle$, we have

$$
\begin{aligned}
\rho_{A} &= \text{tr}_{R} |\Phi\rangle\langle\Phi| = \frac{1}{2} \left(|ge\rangle+|eg\rangle\right) \left(\langle ge|+\langle eg|\right) \\
&=\frac{1}{2}\left(|ge\rangle\langle ge| + |eg\rangle\langle eg| + |ge\rangle\langle eg| + |ge\rangle\langle ge| \right)
\end{aligned}
$$

which is a pure state and perfectly coherent.

In conclusion, we found that the off-diagonal elements of the reduced density operator is sensitive to the entanglement and indeed a good indicator of coherency.


## Methods of open quantum system

One of the objective of the present lecture note is to calculate the reduced density matrix for the emitters by tracing out the radiation degrees of freedom.  How can we calculate the reduced density of a subsystem?  Do we have to compute the while system and take a partial trace?  That needs substantial mathematical and computational power. We cannot use the traditional quantum mechanics for a subsystem since it is developed only for a closed system.  

Fortunately, a method of open quantum system has been developed and widely used in quantum optics.  Although it is an approximated method and valid only in certain situation, there are many evidences that the method works quite well for quantum optics The method is widely used.  The method will be introduced in {numref}`ch:openqm-qme`.
