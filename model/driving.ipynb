# External driving

Optionally, we add an external driving which induces transition between $|g\rangle$ and $|e\rangle$.  If a coherent laser beam is used, the driving of multiple atoms is synchronized and creates coherency among atoms. Interesting interference can be observed from such coherency, which we will discuss later in the chapters of case studies.

## Coupling Hamltonian

The photons in an ideal laser beam are in a coherent state (Glauber state) and the corresponding radiation field behaves almost like a classical field.  It is a reasonable approximation to model the driving using a classical field:

$$
\mathbf{E}(\mathbf{x},t) = \mathbf{E}_{L} \cos(\mathbf{k}_{L}\cdot \mathbf{x} -  \omega_{L} t) = \frac{1}{2} \left(  \mathbf{E}^{(+)}_{L} e^{-i (\mathbf{k}_{L}\cdot \mathbf{x} -  \omega_{L} t)} + \mathbf{E}^{(-)}_{L}(\mathbf{x}) e^{i (\mathbf{k}_{L}\cdot \mathbf{x} -  \omega_{L} t)} \right)
$$

where $\mathbf{E}_{L}$, $\mathbf{k}_{L}$, and $\omega_{L}=c \mathbf{k}_{L}$ are field strength, wave vector, and frequency of the driving laser beam.

The coupling Hamiltonian with the atoms located at $\mathbf{x}_{j}$ is given by

$$
\begin{aligned}
H_{L}(t) &= -  \sum_{j} \mathbf{d}_{j} \cdot \mathbf{E}(\mathbf{x}_j,t) \\
&= - \frac{1}{2}\sum_{j}
\left[\Omega e^{i (\mathbf{k}_{L}\cdot \mathbf{x}_{j} -  \omega_{L} t)} + \tilde{\Omega} e^{-i (\mathbf{k}_{L}\cdot \mathbf{x}_{j} -  \omega_{L} t)}\right] \sigma^{+}_{j} 
-\frac{1}{2} \sum_{j}\left[\Omega^{*} e^{-i (\mathbf{k}_{L}\cdot \mathbf{x}_{j} -  \omega_{L} t)} + \tilde{\Omega}^{*} e^{i (\mathbf{k}_{L}\cdot \mathbf{x}_{j} -  \omega_{L} t)}\right] \sigma^{-}_{j} 
\end{aligned}
$$(eq:H_D)

where $\Omega \equiv \mathbf{d}\cdot \mathbf{E}^{(-)}_{L}$ is the Rabi frequency and $\tilde{\Omega} \equiv \mathbf{d}^{*} \cdot \mathbf{E}^{(-)}_{L}$ the counter-rotating frequency.   

As we discussed in {numref}`ch:radiation-field`, this interaction Hamiltonian contains terms that do not conserve energy. We want to get rid of it.  In addition, the driving Hamiltonian is time dependent, which makes theoretical analysis of time evolution a little complicated. Although we can deal with it using the [Floquet theory](https://en.wikipedia.org/wiki/Floquet_theory), we want to avoid it if possible. A reasonable approximation called [rotating wave approximation](https://en.wikipedia.org/wiki/Rotating-wave_approximation) (RWA) removes the time dependency.

## Rotating wave approximation

The approximation takes the two steps.  The first step remove the non-energy conserving terms.  In the secod step, tie-dependency is removed.

To remove unwanted terms, we write $H_{L}$ in the interaction picture.

$$
\begin{aligned}
H^{\prime}_{L} = U H_{L} U^{\dagger} =  &- \frac{1}{2} \sum_{j}
\left[\Omega e^{i \mathbf{k}_{L}\cdot \mathbf{x}_{j}} e^{i(\omega_{0}-\omega_{L}) t} + \tilde{\Omega} e^{-i \mathbf{k}_{L}\cdot \mathbf{x}_{j}} e^{i (\omega_{0}+ \omega_{L}) t}\right] \sigma^{+}_{j} \\
&-\frac{1}{2} \sum_{j}\left[\Omega^{*} e^{-i \mathbf{k}_{L}\cdot \mathbf{x}_{j}} e^{-i (\omega_{0}-\omega_{L}) t} + \tilde{\Omega}^{*} e^{i \mathbf{k}_{L}\cdot \mathbf{x}_{j}} e^{-i(\omega_{0}+\omega_{L} t)}\right] \sigma^{-}_{j} 
\end{aligned}
$$(eq:H_L_int)

where the unitary transformation is given by

$$
U = e^{-i H_{0} t} = e^{-i \omega_0 t/2} |g_{j}\rangle\langle g_{j}| + e^{i \omega_0 t/2} |e_{j}\rangle\langle e_{j}|.
$$

where $|g_j\rangle$ and $|e_j\rangle$ are two states of the $j$-th atom.

Assuming perfect or near resonance condition,  $\Delta = \omega_{0} - \omega_{L} \approx 0$, $e^{i (\omega_{0} + \omega_{L})t}$ oscillates much faster than $e^{ i (\omega_{0} - \omega_{L})t}$.  We omit the fast oscillating term since they will be canceled out.  The interacting Hamiltonian is simplified to

$$
(H^{\text{RWA}}_{L})^\prime = - \frac{1}{2}\sum_{j}
\left[\Omega e^{i \mathbf{k}_{L}\cdot \mathbf{x}_{j}} e^{i\Delta t} \sigma^{+}_{j}
+ \Omega^{*} e^{-i \mathbf{k}_{L}\cdot \mathbf{x}_{j}} e^{-i \Delta t} \sigma^{-}_{j} \right]
$$

Moving back to the Schroedinger picture, the coupling becomes

$$
H^{\text{RWA}}_{L}(t) = -\frac{1}{2}\sum_{j}
\left[\Omega e^{i \mathbf{k}_{L}\cdot \mathbf{x}_{j}} e^{-i\omega_{L} t} \sigma^{+}_{j}
+ \Omega^{*} e^{-i \mathbf{k}_{L}\cdot \mathbf{x}_{j}} e^{i\omega_{L} t} \sigma^{-}_{j} \right]
$$

Now, the non-energy-conserving term is gone.  However, it still explicitly depends on time.  Next we shift the origin of energy. WE can measure energy from any reference we like.  We apply  another unitary transformation by $V = e^{ i \sum_{j} \frac{\omega_{L}}{2}\sigma^{z}_j}$ which shift all energies by $-\omega_{L}$.  The free energy of atoms becomes

$$
H^\text{RWA}_{0} = V H_{0} V^{\dagger} = \sum_{j} \frac{\Delta}{2} \sigma^{z}_{j}
$$

and the coupling Hamiltonian

$$
\begin{aligned}
H^\text{RWA}_{L} &=V  H^\text{RWA}_{L}(t) V^{\dagger} \\
&= - \frac{1}{2}\sum_{j}
\left(\Omega e^{i \mathbf{k}_{L}\cdot \mathbf{x}_{j}} \sigma^{+}_{j}
+ \Omega^{*} e^{-i \mathbf{k}_{L}\cdot \mathbf{x}_{j}} \sigma^{-}_{j} \right) \\
& = - \frac{|\Omega|}{2} \sum_{j} \left (e^{i \phi_{j}} \sigma^{+}_{j} + e^{-i \phi_{j}} \sigma^{-}_{j}\right)
\end{aligned}
$$(eq:H_L_RWA)

where $\phi_{j} = \mathbf{k}_{L}\cdot \mathbf{x}_{j} + \text{phase of }\Omega$.  This phase plays some crucial roles in some cases.

The Hamiltonian of the driven system with RWA is 

$$
H^\text{RWA}_{S} = H^\text{RWA}_{0} +  H^\text{RWA}_{L}  = \sum_{j} \left[ \frac{\Delta}{2} \sigma^{z}_{j} - \frac{|\Omega|}{2} \left (e^{i \phi_{j}} \sigma^{+}_{j} + e^{-i \phi_{j}} \sigma^{-}_{j}\right) \right].
$$

where $\Delta = \omega_{0}-\omega_{L}$ is the detuning of the driving.  If the tuning is perfect ($\Delta = 0$), then we have only the driving term in the system Hamiltonian.
