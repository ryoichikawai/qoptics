# Phonon bath

When the emitters are embedded in a solid, they interact with phonons.  The collision between the emitters and phonons results in their entanglement and in turn it causes decoherence.  Such collisions induce a type of decoherence known as dephasing. (See {numref}`ch:coherence`).  Furthermore, the emitters exchange phonons generating efffetive interaction between them, similar to the dipole-diplole coupling, which is already discussed in {numref}`sec:emitter`.

## Pure dephasing by phonon scattering

The elastic scattering of phonon by an emitter does not cause transition in the emitter itself. Hence, no energy exchange between the phonon and the emitter.  The interaction Hamiltonian is given by

$$
H_{\phi} = \sum_{j} \sum_{\mathbf{q}} \left( g_\mathbf{q} e^{i \mathbf{q}\cdot \mathbf{x}_{j}} b_\mathbf{q} + g_\mathbf{q}^{*} e^{-i \mathbf{q}\cdot \mathbf{x}_{j}} b^{\dagger}_\mathbf{q} \right) \sigma^{z}_{j}
$$

where $b_\mathbf{a}$ and $b^{\dagger}_\mathbf{a}$ are respectively annhihilation and creation operator of a phonon with momentum $\mathbf{q}$, and $g_\mathbf{q}$ is a coupling constant.  This form of interaction causes pure dephasing. {cite:p}`ContrerasPulido2008,Hu2011,Rice2025`  

## Collective dephasing

The emitters coupled to phonon collectively when the distance between the emitters is shorter than the wave length of phonon. This results in superradiance and subradiance of phonon.  Under the subradiance condition, a special entanglement is formed which protects the emitters from dephasing. {cite:p}`Zanardi1997,Lidar1998` 

## Phonon-mediated coupling

The phonon-mediate coupling between the emitters (quantum dots) is expressed as

$$
H_{\phi-\phi} = \sum_{i\ne j} \Gamma(\mathbf{x}_{i} - \mathbf{x}_{j}) \sigma^{z}_{i}\sigma^{z}_{j}
$$

where the coupling strength is given by

$$
\Gamma(\mathbf{x}) = \sum_{\mathbf{q}} \frac{2|g_\mathbf{q}|^2}{\omega_{\mathbf{q}}} \cos(\mathbf{q}\cdot\mathbf{x})
$$

where $\omega_\mathbf{q}$ is the energy of phonon with momentum $\mathbf{q}$.
