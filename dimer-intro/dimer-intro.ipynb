# A pair of emitters

We investigates the coherency properties of radiation emitted by a pair (dimer) of emitters. As we discussed in {numref}`ch:coherence`, coherency can be lost if we look at only one subspace of a bipartite system depending on the presence of entanglement between the subspace and the rest of the whole space.  

The present system consists of a system of emitters and a radiation field. When we discuss coherence in the emitters, we look at the off-diagonal elements of the reduced density in a certain basis.  On the hand hand, the coherence of the radiation field is measured by coherence functions.  The two sides of coherency are of course related through the entanglement.

The key parameter that controls the coherency is the distance between the two emitters relative to the wave length of light they emit.  The vector from emitter 1 to emitter 2 is denoted as $\mathbf{r}_{12} = \mathbf{x}_{2} - \mathbf{x}_{1}$ and its magnitude is $r_{12}$. The wave length is denoted as $\lambda_{0} = 2\pi/k_{0} = 2 \pi c/\omega_{0}$.  Then, the key parameter is $r_{12}/\lambda_{0} = \frac{1}{2\pi} k_{0} r_{12}$. See {numref}`ssec:r12` for the typical parameter values for quantum dots. We will quantitatively investigate the coherence with respect to the parameter.

We compute the reduced density of the emitters using the quantum master equation introduced in {numref}`ch:openqm-qme`. The coherence functions are evaluated based using the far field modeled by Glauber's detection theory.  Furthermore, we unravel the dynamics using the jump method introduced in {numref}`ch:openqm-jump`.  A summary of method will be given in next section.

We discuss three cases in details.

1. **Thermal fluorescence from an incoherent pair**  ($r_{12}/\lambda_{0} \gg 1$)
2. **Thermal fluorescence from a coherent pair: Superradiance** ($r_{12}/\lambda_{0} \ll 1$)
3. **Coherence by driving: Interference and bunching/antibunching** ($r_{12}/\lambda_{0} \gg 1$)
