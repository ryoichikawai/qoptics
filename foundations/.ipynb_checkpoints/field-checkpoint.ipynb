# Photon gas and Radiation field

## Photon gas

The quantum dots are surrounded by certain environments.  One is a photon gas and another popular one is phonon gas.  We will discuss the phonon environment later.  Here we focus on the photon gas.  The free Hamiltonian of photon is  defined by

$$
H_{R} = \sum_{\ell} \omega_{\ell} \left (a^{\dagger}_{\ell} a_{\ell} + \frac{1}{2} \right)
$$

where $a_{\ell}$ and $a_{\ell}^{\dagger}$ are annihilation and creation operators of the $\ell$-th mode with momentum $\mathbf{k}_{\ell}$ respectively.  We will omit the zero point energy in the rest of this note.

Initially the photon gas is assumed to be in a thermal equilibrium at temperature $T$.  The average occupation number follows the Planck distribution (Boltzman constant $k_{B}=1$ is assumed)

$$
\langle a_{\ell}^{\dagger} a_{\ell}\rangle = \frac{1}{e^{\omega_{\ell}/T} - 1}.
$$

## Radiation fields

Since the radiation field is in a free space, it is natural to expand the field with the plane wave basis set. Then, the electric field at position $\mathbf{x}$ and time $t$ is given by

$$
\mathbf{E}(\mathbf{x}) = \sum_{\ell} \mathbf{e}_{\ell} \left( \mathcal{E}_{\ell} a_{\ell} e^{i \mathbf{k}_{\ell} \cdot \mathbf{x}} +  \mathcal{E}^{*}_{\ell} a^{\dagger}_{\ell} e^{-i \mathbf{k}_{\ell} \cdot \mathbf{x}}\right )
$$(eq:qfield-planewave)

where $\mathcal{E}_{\ell} = i \sqrt{\frac{\hbar \omega_{\ell}}{2 \epsilon_{0} V}}$ and $\epsilon_{\ell}$ is the polarization vector for the $\ell$-th mode.

We also use another popular notation which splits the field to the two parts:

$$
\mathbf{E}^{(+)}(\mathbf{x}) = \sum_{\ell} \mathbf{e}_{\ell} \mathcal{E}_{\ell} a_{\ell} e^{i \mathbf{k}_{\ell} \cdot \mathbf{x}}, \qquad
\mathbf{E}^{(-)}(\mathbf{x}) = \sum_{\ell} \mathbf{e}_{\ell} \mathcal{E}^{*}_{\ell} a^{\dagger}_{\ell} e^{-i \mathbf{k}_{\ell} \cdot \mathbf{x}}.
$$

Remember that the electric field $\mathbf{E}$ is a quantum mechanical operator.  If the Heisenberg picture is used, $a$ and $a^\dagger$ are time-dependent and thus $\mathbf{E}$ is time-dependent.  If the Schrödinger picture is used, then $\mathbf{E}$ does not depend on time.  We will choose a convenient picture depending on the situation.

Another remark is that the expectation value of the field vanishes ($\langle \mathbf{E} \rangle$=0). We are interested in expectation value $\langle \mathbf{E}^{(-)} \cdot\mathbf{E}^{(+)}\rangle$ which  corresponds to the intensity of light.
