(sec:coherence-g1)=
# First-order coherence function

## Definition

The first-order coherence function is defined as correlation of the field amplitude at two different spacetime points: 

$$
g^{(1)}(t_{1},t_{2};\mathbf{x}_1,\mathbf{x}_2) = 
\frac{\langle\mathbf{E}^{(-)}(\mathbf{x}_{1},t_{1}) \cdot \mathbf{E}^{(+)}(\mathbf{x}_2,t_{2})\rangle}
{\sqrt{\langle\mathbf{E}^{(-)}(\mathbf{x}_{1},t_{1})\cdot\mathbf{E}^{(+)}(\mathbf{x}_{1},t_{1})\rangle \langle\mathbf{E}^{(-)}(\mathbf{x}_{2},t_{2})\cdot\mathbf{E}^{(+)}(\mathbf{x}_{2},t_{2})\rangle}}
$$(eq:g1-full)

where $(\mathbf{x}_1,t_{1})$ and $(\mathbf{x}_2,t_{2})$ are two "measurement" points.  This function is complex. The magnitude of $g^{(1)}$, known as the complex degree of coherence, is bound between $0$ and $0$. A value of $1$ indicates perfect coherence, while $0$ means the fields are completely uncorrelated.

We will focus on its special case, temporal correlation measured at a single location:

$$
g^{(1)}(\tau;\mathbf{x}) = 
\frac{\langle\mathbf{E}^{(-)}(\mathbf{x},t) \cdot \mathbf{E}^{(+)}(\mathbf{x},t+\tau)\rangle}
{\langle\mathbf{E}^{(-)}(\mathbf{x},t)\cdot\mathbf{E}^{(+)}(\mathbf{x},t)\rangle}.
$$(eq:g1-temp)

which measures _coherence time_ (how long the radiation field remembers its phase.)

$$
\lim_{\tau\rightarrow\infty} g^{(1)}(\tau;\mathbf{x}) = 0
$$

and 

$$
g^{(1)}(0;\mathbf{x}) = 1
$$

at any position.

## Quantum expression

In quantum oprics, it is more convenient to write Eq. {eq}`eq:g1-temp` with photon operators. Substituting {eq}`eq:quantum-free-field-pm`, Eq. {eq}`eq:g1-temp` is expressed as

$$
g^{(1)}(\tau;\mathbf{x}) = \frac{\sum_{i,j} f_{ij}(\mathbf{x}) \langle a^{\dagger}_{i}(t+\tau) a_{j}(t)\rangle}{\sum_{i,j} f_{ij}(\mathbf{x})  \langle a^{\dagger}_{i}(t) a_{j}(t)\rangle}
$$(eq:g1-quantum)

where

$$
f_{ij}(\mathbf{x}) = \mathbf{e}_{i}\cdot \mathbf{e}_{j}\, \mathcal{E}_{i} \mathcal{E}^{*}_{j} \, e^{i (\mathbf{k}_{i}-\mathbf{k}_{j})\cdot \mathbf{x}}.
$$

To interpret Eq. {eq}`eq:g1-quantum` we rewrite it in slightly different expression:

$$
g^{(1)}(\tau,\mathbf{x}) = \sum_{ij} w_{ij}(\mathbf{x}) g^{(1)}_{ij}(\tau).
$$(eq:g1-weighted-sum)

where 

$$
g^{(1)}_{ij}(\tau) = \frac{\langle a^{\dagger}_{i}(t+\tau)a_{j}(t)\rangle_t}{\langle a_{i}^{*}(t)a_{j}(t)\rangle}
$$(eq:g1-temp-only)

and weight functions

$$
w_{ij}(\mathbf{x}) = \frac{f_{ij}(\mathbf{x},\mathbf{x})  \langle a^{\dagger}_{i}(t) a_{j}(t)\rangle}
{\sum_{ij}f_{ij}(\mathbf{x},\mathbf{x})  \langle a^{\dagger}_{i}(t) a_{j}(t)}.
$$(eq:wij)

The weight is normalized as $\sum_{ij} w_{ij}(\mathbf{x}) = 1$.  Equation {eq}`eq:g1-weighted-sum` indicates that at each oint $\mathbf{x}$, the contribution of temporal correlation functions $g^{(1)}_{ij}$ is different.

## Is $g^{(1)}$ measurable?

The first-order coherence function is a correlation of electric fields measured at two spacetime points. Is that something we can measure in a real world experiment?  It is not. In quantum mechanics, electric field is not an observable. It is even not a self-adjoint operator and thus $g^{(1)}(\tau)$ is inherently complex. It is like measuring a wave function or a density operator. We know that neither is possible. Based on the principle of quantum mechanics, there is no way to measure $g^{(1)}(\tau)$! 

Then, why do we use it? It is a mathematical utility to investigate coherence. In fact, it is a powerful tool. Is there anyway to reconstruct it from experimental data?  Yes, there is. It is called [quantum tomography](https://en.wikipedia.org/wiki/Quantum_tomography).  You measure many different cases and if there is enough information in data, it is possible to _infer_ the function by fitting data to the theory. Essentially, we adjust parameters until all data can be fit to the theory.  It is a tedious process.

There is no such difficulty with the second-order coherence function since it is a correlation of intensities which can be measured at any time and any point.  It is also noted that quantum trajectory measures also intensity.  Thus we can evaluate $g^{(2)}(\tau)$ from the quantum trajectories but not $g^{(1)}(\tau)$.
