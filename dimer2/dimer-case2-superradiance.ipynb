```python
import matplotlib.pyplot as plt
import numpy as np
from qutip import *

# Parameters
omega0 = 1.0   # detuning
gamma0 = 0.1  # spontaneous emission rate 
temperature = 1.0  # tempeature of photon gas
NT = 1/(np.exp(omega0/temperature)-1) # Planck distribution
gamma_a = gamma0 * NT   # coefficient to absorption
gamma_e = gamma0 * (NT+1)  # coefficient to emission
Omega12 = 10*gamma0

phi = np.pi/4

i2 = qeye(2)
sz = [tensor(sigmaz(),i2),tensor(i2,sigmaz())]
sp = [tensor(sigmap(),i2),tensor(i2,sigmap())]
sm = [tensor(sigmam(),i2),tensor(i2,sigmam())]
ep = [(sp[0]+sp[1])/np.sqrt(2),(sp[0]-sp[1])/np.sqrt(2)]
em = [(sm[0]+sm[1])/np.sqrt(2),(sm[0]-sm[1])/np.sqrt(2)]

# Hamiltonian
H = 0.5*omega0*(sz[0]+sz[1]) + Omega12*(sp[0]*sm[1]+sp[1]*sm[0])

# Collapse operators
c_ops = [np.sqrt(gamma_e)*em[0], np.sqrt(gamma_e)*em[1],np.sqrt(gamma_a)*ep[0], np.sqrt(gamma_a)*ep[1]]

# Steady state
rho_p = steadystate(H,c_ops)

# product basis
u=[]
u.append(tensor(basis(2,0),basis(2,0)))
u.append(tensor(basis(2,1),basis(2,0)))
u.append(tensor(basis(2,0),basis(2,1)))
u.append(tensor(basis(2,1),basis(2,1)))

# Dicke basis
v=[]
v.append(u[0])
v.append((u[1]+u[2])/np.sqrt(2))
v.append((u[1]-u[2])/np.sqrt(2))
v.append(u[3])

# change of the basis set_
rho_s = np.zeros((4, 4), dtype=complex)
for i in range(4):
    for j in range(4):
        rho_s[[i],[j]] = rho_p.matrix_element(v[i],v[j])

print("** Diagonal elements of density operator **")
print("       numerical   theory")
print("ρ_gg  = {0:7.4f}".format(rho_s[0,0].real))
print("ρ_ss  = {0:7.4f}".format(rho_s[1,1].real))
print("ρ_aa  = {0:7.4f}".format(rho_s[2,2].real))
print("ρ_ee  = {0:7.4f}".format(rho_s[3,3].real))
print("ρ_sa  = {0:7.4f}".format(rho_s[1,2]))
```
