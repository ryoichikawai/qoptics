# The Model

The free Hamiltonian of the emitters, $H_{0}$, remains the same as Eq. {eq}`eq:dimer-H0`. 
It has the eigen spectra $E_{ee}=\omega_{0}$, $E_{ge}=E_{eg}=0$, and $E_{gg}=-\omega_{0}$.  Notice the degeneracy. The energy eigen vectors are not unique. Any linear combination of $|ge\range$ and $|eg\rangle$ is an energy eigen state.  However, when $\Gamma_{12}$ vanishes, the jump operators $\sigma_^{\pm}_{j}$ in the dissipator pick $|eg\rangle$ and $|ge\rangle$ as preferred basis kets. 

In this chapter we invetigate the opposite case.  With $\Gamma_{12}$ and also the dipole-dipole coupling, a different basis set is chosen by the radiation field.

## The dipole-dipole coupling

When the dipole=dipole coupling {eq}`eq:H_dd` is added, the degeneracy is lifted.  The new energy eigenvectors are conveniently  written with the angular momentum eigenstates $|J,M\rangle$.  The triplet states  $|1,1\rangle = |ee\rangle$, $|1,0\rangle = (|eg\rangle+|ge\rangle)/\sqrt{2}$, and $|1,-1\rangle = |gg\rangle$ and the singlet state $|0,0\rangle = |eg\rangle-|ge\rangle)/\sqrt{2}$.  The corresponding energy eigenvales are $E_{1,1}=\omega_{0}$, $E_{1,0}=\Omega_{12}$, $E_{1,-1}=-\omega_0$ and $E_{0,0}=-\Omega_{12}$, respectively. The transitions among the four states generate two different frequency of light, $\omega_{\pm} = \omega_0 \pm \Omega_{12}$.





In the previous case, the spontaneous emission and absorption took place at random via the collapse operators $\sigma^{\pm}_{j}$.  In the present case, the emission and absorption still happen randomly but by the collapse operators    This energy eigenstates induce coherency between.

A problem is that $\Omega_{12}$ diverges as $r_{12} \rightarrow 0$. The unrealistic assumption that the emitter is a point particle causes this divergence.  Assuming that the charge is Gaussian distributed with radius $s$, the coefficient near $r_{12}=0$ is given by 

$$
\Omega_{12} (r_{12} \ll \lambda_{0}) = \frac{3\gamma_{0}}{4} (1-3 \cos^{2}\theta_{12}) \frac{1}{(k_{0}r_{12})^2} \left[\text{erf}\left(\frac{r_{12}}{s}\right) - \sqrt{\frac{2}{\pi}} \left(\frac{r_{12}}{s}\right) e^{-\frac{1}{2}\left(r_{12}/s
\right)^2}\right]
$$

which is finite even at $r_{12}=0$.   However, $r_{12}<s$ does not have physical reality since two emitters overlap.  We treat $\Omega_{12}$ as a parameter ranging from $\gamma_{0}$ to $100 \gamma_{0}$.
