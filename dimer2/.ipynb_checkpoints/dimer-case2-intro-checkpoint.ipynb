# Dimer II: Coherence by indistinguishability

In the previous chapter, we considered the case where $\Gamma_{12}=0$. In this chapter, we consider the opposite limit, $\Gamma_{12} \rightarow \gamma_{0}$ where the distance between the two emitters is too small ($k_{0}r_{12} \ll 1$) for the far field to distinguish them.  The dipole-dipole coupling is rather strong and thus the degeneracy in the energy eigenvalues is lifted.

In the extreme limit $\Gamma_{12}=\gamma_{0}$, a dark state appears and superradiance is observed.
